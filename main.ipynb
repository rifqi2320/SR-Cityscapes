```python
"""
Cityscapes semantic‑segmentation evaluation pipeline (patch‑focused with SR comparison).

This script evaluates a DeeplabV3+ model on Cityscapes *and* provides rich, patch‑centric
visual analytics:

1. **Mask evolution** – for the N largest small‑object patches, view overlays of
   *Raw → High‑confidence → Refined* predictions.
2. **Super‑resolution quality** – side‑by‑side crops of each patch *before* and
   *after* Swin2SR upscaling.
3. **Patch‑level mask gains** – compare segmentation masks *before* vs *after*
   super‑resolution inference, overlayed on each patch crop.
4. Console metrics for IoU and misclassification deltas.

Adjust everything from the **CONFIG** block.

Author: <your‑name>
Date: 2025‑05‑20 (updated 2025‑05‑21)
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"          # Cityscapes root directory
CHECKPOINT_PATH: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"  # Trained weights
CHECKPOINT_PATH_SPLICE: str = "./checkpoints_sr_splice/best.pth"  # Trained weights
OUTPUT_STRIDE: int = 16                               # DeeplabV3+ output stride
NUM_CLASSES: int = 19                                 # Cityscapes train IDs
SMALL_OBJECT_CLASSES: list[int] = [6, 7]              # Classes to super‑resolve/patch (bus, train)
PRED_CONF_THRESHOLD: float = 0.5                      # Confidence threshold
MORPH_KERNEL_SIZE: int = 20                           # Closing kernel size
MIN_INSTANCE_SIDE: int = 50                           # Minimum w/h of patch (px)
SR_MODEL_NAME: str = "caidas/swin2SR-lightweight-x2-64"  # Super‑resolution model
SAMPLE_INDEX: int = 10                                 # Validation image index
VISUALISE_TOP_N_PATCHES: int = 4                      # How many largest patches to show

# Choose device (CPU fallback)
import torch
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================
# LIBRARIES & MODEL LOADING
# ============================
from typing import List, Tuple
import numpy as np
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms as T
from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes
import matplotlib.patches as mpatches

# ----------------------------
# Helper dicts
# ----------------------------
CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# ============================
# UTILITY FUNCTIONS
# ============================

def calculate_iou(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> np.ndarray:
    """Class‑wise IoU (ignores absent classes)."""
    ious: List[float] = []
    for cls in range(num_classes):
        inter = np.logical_and(pred == cls, true == cls).sum()
        union = np.logical_or(pred == cls, true == cls).sum()
        if union > 0:
            ious.append(inter / union)
    return np.asarray(ious, dtype=float)


def closing(mask: np.ndarray, kernel: int = MORPH_KERNEL_SIZE) -> np.ndarray:
    k = np.ones((kernel, kernel), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_CLOSE, k)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    px, py = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - px), max(0, y - py)
    x2, y2 = min(img.shape[1], x + w + px), min(img.shape[0], y + h + py)
    return img[y1:y2, x1:x2]

# ----------------------------
# Misclassification helpers
# ----------------------------

def compare_predictions(gt: np.ndarray, raw: np.ndarray, refined: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    mis_raw = raw != gt
    mis_ref = refined != gt
    improved = np.logical_and(mis_raw, ~mis_ref)
    worsened = np.logical_and(~mis_raw, mis_ref)
    return improved, worsened

# ============================
# VISUALISATION HELPERS
# ============================

def _rank_patches_by_area(instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]]) -> List[int]:
    areas = [(idx, bbox[2] * bbox[3]) for idx, (*_, bbox) in enumerate(instances)]
    areas.sort(key=lambda t: -t[1])
    return [idx for idx, _ in areas]


def show_patch_triplets(img: np.ndarray,
                        raw: np.ndarray,
                        high: np.ndarray,
                        refined: np.ndarray,
                        instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]],
                        top_n: int = 4):
    """Overlay mask evolution for top‑n patches."""
    if not instances:
        return
    idxs = _rank_patches_by_area(instances)[:top_n]
    rows = len(idxs)
    fig, axes = plt.subplots(rows, 3, figsize=(9, 3 * rows))
    if rows == 1:
        axes = np.expand_dims(axes, 0)
    titles = ["Raw", "High‑conf", "Refined"]
    for r, i in enumerate(idxs):
        _, _, (x, y, w, h) = instances[i]
        patch_img = img[y:y+h, x:x+w]
        masks = [raw[y:y+h, x:x+w], high[y:y+h, x:x+w], refined[y:y+h, x:x+w]]
        for c, m in enumerate(masks):
            ax = axes[r, c]
            ax.imshow(patch_img)
            ax.imshow(m, alpha=0.5, cmap="jet")
            ax.axis("off")
            if r == 0:
                ax.set_title(titles[c])
    # legend
    patches = []
    for cls in range(NUM_CLASSES):
        label = CITYSCAPES_LABELS.get(ID2TRAINID.get(cls, str(cls)), str(cls))
        color = plt.cm.jet(cls / NUM_CLASSES)
        patches.append(mpatches.Patch(color=color, label=f"{label} ({cls})"))
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 0.5),
               loc='center left', fontsize=8,
               borderaxespad=0.)
    plt.suptitle("Patch evolution (raw → high‑conf → refined)")
    plt.subplots_adjust(top=0.9)
    plt.tight_layout()
    plt.show()


def show_sr_pairs(originals: List[np.ndarray], sr_imgs: List[np.ndarray], top_n: int = 4):
    """Display original vs super‑resolved crops for top‑n patches."""
    if not originals:
        return
    rows = min(top_n, len(originals))
    fig, axes = plt.subplots(rows, 2, figsize=(6, 3 * rows))
    if rows == 1:
        axes = np.expand_dims(axes, 0)
    for r in range(rows):
        axes[r, 0].imshow(originals[r])
        axes[r, 0].set_title("Original")
        axes[r, 0].axis("off")
        axes[r, 1].imshow(sr_imgs[r])
        axes[r, 1].set_title("Super‑resolved")
        axes[r, 1].axis("off")
    plt.tight_layout()
    plt.show()


def show_patch_mask_pairs(crops: List[np.ndarray],
                          preds_before: List[np.ndarray],
                          preds_after: List[np.ndarray],
                          top_n: int = 4):
    """Compare per‑patch segmentation before vs after SR."""
    if not crops or not preds_before or not preds_after:
        return
    rows = min(top_n, len(crops))
    fig, axes = plt.subplots(rows, 2, figsize=(8, 3 * rows))
    if rows == 1:
        axes = np.expand_dims(axes, 0)
    for r in range(rows):
        crop = crops[r]
        pb = preds_before[r]
        pa = preds_after[r]
        # Before
        axes[r, 0].imshow(crop)
        axes[r, 0].imshow(pb, alpha=0.5, cmap="jet")
        axes[r, 0].set_title("Mask • Before SR")
        axes[r, 0].axis("off")
        # After
        axes[r, 1].imshow(crop)
        axes[r, 1].imshow(pa, alpha=0.5, cmap="jet")
        axes[r, 1].set_title("Mask • After SR")
        axes[r, 1].axis("off")
    plt.suptitle("Per‑patch mask comparison (pre vs post SR)")
    plt.tight_layout()
    plt.show()

# ============================
# MAIN PIPELINE
# ============================

def main() -> None:
    # Model ---------------------------------------------------------------
    model = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES,
                                    output_stride=OUTPUT_STRIDE,
                                    pretrained_backbone=False)
    ckpt = torch.load(CHECKPOINT_PATH, map_location=DEVICE, weights_only=False)
    model.load_state_dict(ckpt["model_state"], strict=False)
    model.eval().to(DEVICE)

    # Data ----------------------------------------------------------------
    ds = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
    img_pil, mask_pil = ds[SAMPLE_INDEX]
    img_np = np.array(img_pil)
    mask_np = np.vectorize(ID2TRAINID.get)(np.array(mask_pil))

    # Normalisation transform reused for both full‑image and per‑patch inference
    norm = T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])

    def preprocess(img_np: np.ndarray) -> torch.Tensor:
        return norm(torch.from_numpy(img_np.transpose(2, 0, 1)).float() / 255.).unsqueeze(0).to(DEVICE)

    # Forward (full‑frame) -----------------------------------------------
    with torch.no_grad():
        logits = model(preprocess(img_np))[0]
    logits = (logits - logits.min()) / (logits.max() - logits.min())
    pred_map = logits.cpu().numpy()

    # Predictions ---------------------------------------------------------
    raw_pred = pred_map.argmax(axis=0).astype(np.uint8)
    print(f"Mean IoU (raw)          : {calculate_iou(raw_pred, mask_np).mean():.4f}")

    thr_map = np.where(pred_map > PRED_CONF_THRESHOLD, pred_map, 0.0)
    high_pred = thr_map.argmax(axis=0).astype(np.uint8)
    print(f"Mean IoU (thresholded)  : {calculate_iou(high_pred, mask_np).mean():.4f}")

    closed_pred = closing(high_pred)

    # Patch extraction ----------------------------------------------------
    instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]] = []
    for cid in SMALL_OBJECT_CLASSES:
        mask_cls = (closed_pred == cid).astype(np.uint8)
        contours, _ = cv2.findContours(mask_cls, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE:
                continue
            crop = crop_with_padding(img_np, x, y, w, h, pad=0.5)
            instances.append((cid, crop, (x, y, w, h)))

    # Super‑resolution ----------------------------------------------------
    proc = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
    sr_net = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()

    def sr(img: np.ndarray) -> np.ndarray:
        ins = proc(img, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            out = sr_net(**ins)
        rec = out.reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
        return (np.moveaxis(rec, 0, -1) * 255).round().astype(np.uint8)

    refined_pred = high_pred.copy()
    sr_originals: List[np.ndarray] = []     # original RGB crops
    sr_results: List[np.ndarray] = []       # SR‑enhanced RGB crops
    patch_preds_before: List[np.ndarray] = []
    patch_preds_after: List[np.ndarray] = []

    for cid, crop, (x, y, w, h) in instances:
        # --- 1. Create un‑padded patch for visual alignment ------------
        patch_rgb = img_np[y:y+h, x:x+w]

        # --- 2. Prediction on *original* (padded) crop -----------------
        with torch.no_grad():
            logits_orig = model(preprocess(crop))[0]
        labels_orig = torch.softmax(logits_orig, dim=0).argmax(0).cpu().numpy()
        labels_orig = cv2.resize(labels_orig, (w, h), interpolation=cv2.INTER_NEAREST)

        # --- 3. Super‑resolve + prediction -----------------------------
        sr_crop_full = sr(crop.transpose(2, 0, 1))  # dimensions ~= 2×crop
        # Extract the central (unpadded) region from SR output so it matches bbox
        pad_x, pad_y = int(w * 0.5), int(h * 0.5)
        sr_patch = sr_crop_full[pad_y:pad_y+h*2, pad_x:pad_x+w*2]  # 2× scaling
        # Downscale back to original bbox for overlay consistency
        sr_patch_down = cv2.resize(sr_patch, (w, h), interpolation=cv2.INTER_CUBIC)

        with torch.no_grad():
            logits_sr = model(preprocess(sr_patch_down))[0]
        labels_sr = torch.softmax(logits_sr, dim=0).argmax(0).cpu().numpy()

        # --- 4. Collect visuals ----------------------------------------
        sr_originals.append(patch_rgb)          # unpadded RGB
        sr_results.append(sr_patch_down)        # SR‑enhanced RGB (aligned)
        patch_preds_before.append(labels_orig)  # mask before SR
        patch_preds_after.append(labels_sr)     # mask after SR

        # --- 5. Patch‑wise refined insertion ---------------------------
        refined_pred[y:y+h, x:x+w] = labels_sr

    # Metrics -------------------------------------------------------------
    print(f"Mean IoU (refined)      : {calculate_iou(refined_pred, mask_np).mean():.4f}\n")
    imp, wors = compare_predictions(mask_np, raw_pred, refined_pred)
    print("Misclassification delta:")
    print(f"  Corrected pixels : {imp.sum()}")
    print(f"  New errors       : {wors.sum()}")
    print(f"  Net gain         : {imp.sum() - wors.sum()}")

    # Visuals -------------------------------------------------------------
    show_patch_mask_pairs(sr_originals, patch_preds_before, patch_preds_after, VISUALISE_TOP_N_PATCHES)
    show_patch_triplets(img_np, raw_pred, high_pred, refined_pred, instances, VISUALISE_TOP_N_PATCHES)
    show_sr_pairs(sr_originals, sr_results, VISUALISE_TOP_N_PATCHES)


if __name__ == "__main__":
    main()
```

```python
import cv2

# for every class, do opening and closing
def closing(mask, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_CLOSE, kernel)

def opening(mask, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_OPEN, kernel)

# Apply opening and closing to the predicted mask
opened_mask = closing(pred_map, kernel_size=20)
# opened_mask = opening(closed_mask, kernel_size=20)

# iou before and after
iou_before = calculate_iou(pred_map, mask_np, num_classes=30)
iou_after = calculate_iou(opened_mask, mask_np, num_classes=30)
print(f"Mean IoU before closing: {np.mean(iou_before)}")
print(f"Mean IoU after closing: {np.mean(iou_after)}")

# Display the image and opened mask
show_image_with_mask(image_np, opened_mask, opacity=0.5, title='Cityscapes Image with Opened Mask')

# print the size of all islands for every class
def get_island_sizes(mask):
    # Find contours
    contours, _ = cv2.findContours(mask.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    island_sizes = [cv2.contourArea(c) for c in contours]
    return island_sizes

for class_id in range(1, 30):
    class_mask = (opened_mask == class_id).astype(np.uint8)
    island_sizes = get_island_sizes(class_mask)
    print(f"Class {class_id}: {len(island_sizes)} islands")
    for i, size in enumerate(island_sizes):
        print(f"  Island {i}: {size} pixels")
```

```python
# for every SMALL_OBJECT_CLASSES class, create a cropped area with note of offset, width, and height
# crop the image with the bounding box with padding
def crop_with_padding(image, x, y, w, h, padding=0.1):
    # Calculate padding
    pad_x = int(w * padding)
    pad_y = int(h * padding)
    
    # Calculate new coordinates with padding
    x1 = max(0, x - pad_x)
    y1 = max(0, y - pad_y)
    x2 = min(image.shape[1], x + w + pad_x)
    y2 = min(image.shape[0], y + h + pad_y)
    
    # Crop the image
    cropped_image = image[y1:y2, x1:x2]
    
    return cropped_image

instances = []
for class_id in SMALL_OBJECT_CLASSES:
    class_mask = (opened_mask == class_id).astype(np.uint8)
    # find contours
    contours, _ = cv2.findContours(class_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)

        if w < 50 or h < 50:
            continue

        # crop the image with the bounding box with padding
        cropped_image = crop_with_padding(image_np, x, y, w, h, padding=0.1)
        instances.append((class_id, cropped_image, (x, y, w, h)))


# Display the cropped images
for class_id, cropped_image, bbox in instances:
    plt.imshow(cropped_image)
    plt.title(f"Class {class_id} - BBox: {bbox}")
    plt.axis('off')
    plt.show()
```

```python
from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution

processor = AutoImageProcessor.from_pretrained("caidas/swin2SR-lightweight-x2-64")
sr_model = Swin2SRForImageSuperResolution.from_pretrained("caidas/swin2SR-lightweight-x2-64")

def super_resolve_image(image):
    # Preprocess the image
    inputs = processor(image, return_tensors="pt").to(sr_model.device)

    # Forward pass
    with torch.no_grad():
        outputs = sr_model(**inputs)

    # Post-process the output
    output = outputs.reconstruction.data.squeeze().float().cpu().clamp_(0, 1).numpy()
    output = np.moveaxis(output, source=0, destination=-1)
    output = (output * 255.0).round().astype(np.uint8)  # float32 to uint8
    return output

# Apply super-resolution to the cropped images
sr_images = []
for class_id, cropped_image, bbox in instances:
    sr_image = super_resolve_image(cropped_image.transpose(2, 0, 1))
    sr_images.append((class_id, sr_image, bbox))
# Display the super-resolved images
for class_id, sr_image, bbox in sr_images:
    plt.imshow(sr_image)
    plt.title(f"Super-resolved Class {class_id}")
    plt.axis('off')
    plt.show()
```

```python
"""Cityscapes semantic-segmentation evaluation pipeline (patch-focused with SR comparison).

This script evaluates a DeeplabV3+ model on Cityscapes *and* provides rich, patch-centric
visual analytics:

1. **Mask evolution** – for the N largest small-object patches, view overlays of
   *Raw → High-confidence → Refined* predictions.
2. **Super-resolution quality** – side-by-side crops of each patch *before* and
   *after* Swin2SR upscaling.
3. Console metrics for IoU and misclassification deltas.

Adjust everything from the **CONFIG** block.

Author: <your-name>
Date: 2025-05-20
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"          # Cityscapes root directory
CHECKPOINT_PATH: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"  # Trained weights
CHECKPOINT_PATH_SPLICE: str = "./checkpoints_sr_splice/last.pth"  # Trained weights
OUTPUT_STRIDE: int = 16                               # DeeplabV3+ output stride
NUM_CLASSES: int = 19                                 # Cityscapes train IDs
SMALL_OBJECT_CLASSES: list[int] = [6, 7]              # Classes to super-resolve/patch (bus, train)
PRED_CONF_THRESHOLD: float = 0.5                      # Confidence threshold
MORPH_KERNEL_SIZE: int = 20                           # Closing kernel size
MIN_INSTANCE_SIDE: int = 50                           # Minimum w/h of patch (px)
SR_MODEL_NAME: str = "caidas/swin2SR-lightweight-x2-64"  # Super-resolution model
SAMPLE_INDEX: int = 3                                 # Validation image index
VISUALISE_TOP_N_PATCHES: int = 4                      # How many largest patches to show

# Choose device (CPU fallback)
import torch
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================
# LIBRARIES & MODEL LOADING
# ============================
from typing import List, Tuple
import numpy as np
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms as T
from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes
import matplotlib.patches as mpatches

# ----------------------------
# Helper dicts
# ----------------------------
CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}
TRAINID2COLOR = {c.train_id: c.color for c in DeeplabCityscapes.classes}

# ============================
# UTILITY FUNCTIONS
# ============================

def calculate_iou(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> np.ndarray:
    """Class-wise IoU (ignores absent classes)."""
    ious: List[float] = []
    for cls in range(num_classes):
        inter = np.logical_and(pred == cls, true == cls).sum()
        union = np.logical_or(pred == cls, true == cls).sum()
        if union > 0:
            ious.append(inter / union)
    return np.asarray(ious, dtype=float)


def closing(mask: np.ndarray, kernel: int = MORPH_KERNEL_SIZE) -> np.ndarray:
    k = np.ones((kernel, kernel), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_CLOSE, k)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    px, py = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - px), max(0, y - py)
    x2, y2 = min(img.shape[1], x + w + px), min(img.shape[0], y + h + py)
    return img[y1:y2, x1:x2]

# ----------------------------
# Misclassification helpers
# ----------------------------

def compare_predictions(gt: np.ndarray, raw: np.ndarray, refined: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    mis_raw = raw != gt
    mis_ref = refined != gt
    improved = np.logical_and(mis_raw, ~mis_ref)
    worsened = np.logical_and(~mis_raw, mis_ref)
    return improved, worsened

# ============================
# VISUALISATION HELPERS
# ============================

def _rank_patches_by_area(instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]]) -> List[int]:
    areas = [(idx, bbox[2] * bbox[3]) for idx, (*_, bbox) in enumerate(instances)]
    areas.sort(key=lambda t: -t[1])
    return [idx for idx, _ in areas]


def show_patch_triplets(img: np.ndarray,
                        raw: np.ndarray,
                        high: np.ndarray,
                        refined: np.ndarray,
                        instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]],
                        top_n: int = 4):
    """Overlay mask evolution for top-n patches."""
    if not instances:
        return
    idxs = _rank_patches_by_area(instances)[:top_n]
    rows = len(idxs)
    fig, axes = plt.subplots(rows, 3, figsize=(9, 3 * rows))
    if rows == 1:
        axes = np.expand_dims(axes, 0)
    titles = ["Raw", "High-conf", "Refined"]
    for r, i in enumerate(idxs):
        _, _, (x, y, w, h) = instances[i]
        patch_img = img[y:y+h, x:x+w]
        masks = [raw[y:y+h, x:x+w], high[y:y+h, x:x+w], refined[y:y+h, x:x+w]]
        for c, m in enumerate(masks):
            ax = axes[r, c]
            ax.imshow(patch_img)
            m = np.vectorize(TRAINID2COLOR.get)(m)
            m = np.moveaxis(m, 0, -1)
            ax.imshow(m, alpha=0.5, cmap="jet")
            ax.axis("off")
            if r == 0:
                ax.set_title(titles[c])
    # legend
    patches = []
    for cls in range(NUM_CLASSES):
        if cls not in raw and cls not in high and cls not in refined:
            continue
        lbl = CITYSCAPES_LABELS.get(cls, str(cls))
        color = TRAINID2COLOR.get(cls, (0, 0, 0))
        color = np.array(color) / 255.0
        patches.append(mpatches.Patch(color=color, label=f"{lbl} ({cls})"))
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 0.5), 
               loc='center left', fontsize=8, 
               borderaxespad=0.)
    # show
    plt.suptitle("Patch evolution (raw → high-conf → refined)")
    plt.subplots_adjust(top=0.9)
    plt.tight_layout()
    plt.show()


def show_sr_pairs(originals: List[np.ndarray], sr_imgs: List[np.ndarray], top_n: int = 4):
    """Display original vs super-resolved crops for top-n patches."""
    if not originals:
        return
    rows = min(top_n, len(originals))
    fig, axes = plt.subplots(rows, 2, figsize=(6, 3 * rows))
    if rows == 1:
        axes = np.expand_dims(axes, 0)
    for r in range(rows):
        axes[r, 0].imshow(originals[r])
        axes[r, 0].set_title("Original")
        axes[r, 0].axis("off")
        axes[r, 1].imshow(sr_imgs[r])
        axes[r, 1].set_title("Super-resolved")
        axes[r, 1].axis("off")
    plt.tight_layout()
    plt.show()

# ============================
# MAIN PIPELINE
# ============================

# Model ---------------------------------------------------------------
model = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES,
                                output_stride=OUTPUT_STRIDE,
                                pretrained_backbone=False)
ckpt = torch.load(CHECKPOINT_PATH, map_location=DEVICE, weights_only=False)
model.load_state_dict(ckpt["model_state"], strict=False)
model.eval().to(DEVICE)

model_splice = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES,
                                        output_stride=OUTPUT_STRIDE,
                                        pretrained_backbone=False)
ckpt_splice = torch.load(CHECKPOINT_PATH_SPLICE, map_location=DEVICE, weights_only=False)
model_splice.load_state_dict(ckpt_splice["model_state"], strict=False)
model_splice.eval().to(DEVICE)

# Data ----------------------------------------------------------------
ds = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
img_pil, mask_pil = ds[SAMPLE_INDEX]
img_np = np.array(img_pil)
mask_np = np.vectorize(ID2TRAINID.get)(np.array(mask_pil))

# Forward -------------------------------------------------------------
tensor_inp = T.Compose([
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])(img_pil.convert("RGB")).unsqueeze(0).to(DEVICE)
with torch.no_grad():
    logits = model(tensor_inp)[0]
logits = (logits - logits.min()) / (logits.max() - logits.min())
pred_map = logits.cpu().numpy()

# Predictions ---------------------------------------------------------
raw_pred = pred_map.argmax(axis=0).astype(np.uint8)
print(f"Mean IoU (raw)          : {calculate_iou(raw_pred, mask_np).mean():.4f}")

thr_map = np.where(pred_map > PRED_CONF_THRESHOLD, pred_map, 0.0)
high_pred = thr_map.argmax(axis=0).astype(np.uint8)
print(f"Mean IoU (thresholded)  : {calculate_iou(high_pred, mask_np).mean():.4f}")

closed_pred = closing(high_pred)

# Patch extraction ----------------------------------------------------
instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]] = []
for cid in SMALL_OBJECT_CLASSES:
    mask_cls = (closed_pred == cid).astype(np.uint8)
    contours, _ = cv2.findContours(mask_cls, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE:
            continue
        crop = crop_with_padding(img_np, x, y, w, h, pad=1)
        adjusted_x = int(x - w * 0.5)
        adjusted_y = int(y - h * 0.5)
        adjusted_w = int(w * 1.5)
        adjusted_h = int(h * 1.5)
        # Ensure the adjusted coordinates are within the image bounds
        adjusted_x = max(0, adjusted_x)
        adjusted_y = max(0, adjusted_y)
        adjusted_w = min(adjusted_w, img_np.shape[1] - adjusted_x)
        adjusted_h = min(adjusted_h, img_np.shape[0] - adjusted_y)
        crop = img_np[adjusted_y:adjusted_y+adjusted_h, adjusted_x:adjusted_x+adjusted_w]
        instances.append((cid, crop, (adjusted_x, adjusted_y, adjusted_w, adjusted_h)))

# Super-resolution ----------------------------------------------------
proc = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
sr_net = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()

def sr(img: np.ndarray) -> np.ndarray:
    ins = proc(img, return_tensors="pt").to(DEVICE)
    with torch.no_grad():
        out = sr_net(**ins)
    rec = out.reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
    return (np.moveaxis(rec, 0, -1) * 255).round().astype(np.uint8)

refined_pred = high_pred.copy()
sr_originals: List[np.ndarray] = []
sr_results: List[np.ndarray] = []

comparisons = []

for cid, crop, (x, y, w, h) in instances:
    sr_crop = sr(crop.transpose(2, 0, 1))
    sr_originals.append(crop)
    sr_results.append(sr_crop)

    tc = torch.from_numpy(sr_crop.transpose(2, 0, 1)).float().unsqueeze(0).to(DEVICE)
    unc_tc = torch.from_numpy(crop.transpose(2, 0, 1)).float().unsqueeze(0).to(DEVICE)
    with torch.no_grad():
        logits_c = model_splice(tc)[0]
        unc_logits_c = model_splice(unc_tc)[0]
    probs_c = torch.softmax(logits_c, dim=0).cpu().numpy()
    unc_probs_c = torch.softmax(unc_logits_c, dim=0).cpu().numpy()
    labels_c = probs_c.argmax(axis=0)
    unc_labels_c = unc_probs_c.argmax(axis=0)
    labels_c = cv2.resize(labels_c, (w, h), interpolation=cv2.INTER_NEAREST)
    refined_pred[y:y+h, x:x+w] = labels_c
    comparisons.append((crop, unc_labels_c, labels_c))

# Metrics -------------------------------------------------------------
print(f"Mean IoU (refined)      : {calculate_iou(refined_pred, mask_np).mean():.4f}\n")
imp, wors = compare_predictions(mask_np, raw_pred, refined_pred)
print("Misclassification delta:")
print(f"  Corrected pixels : {imp.sum()}")
print(f"  New errors       : {wors.sum()}")
print(f"  Net gain         : {imp.sum() - wors.sum()}")

# Visuals -------------------------------------------------------------
show_patch_triplets(img_np, raw_pred, high_pred, refined_pred, instances, VISUALISE_TOP_N_PATCHES)
show_sr_pairs(sr_originals, sr_results, VISUALISE_TOP_N_PATCHES)

# show comparison
crop, unc_label, label = comparisons[0]
print(crop.shape, unc_label.shape, label.shape)
# map TRAINID2COLOR to unc_label and label
# TRAINID2COLOR maps int to tuple
patches = []
for cls in range(NUM_CLASSES):
    # only show classes present in the mask
    if cls in unc_label:
        lbl = CITYSCAPES_LABELS.get(cls, str(cls))
        color = TRAINID2COLOR.get(cls, (0, 0, 0))
        # color is 255 based
        color = tuple(c / 255 for c in color)
        patches.append(mpatches.Patch(color=color, label=f"{lbl} ({cls})"))
# add background

patches.append(mpatches.Patch(color=(1, 1, 1), label="Background"))
unc_label = np.vectorize(TRAINID2COLOR.get)(unc_label)
label = np.vectorize(TRAINID2COLOR.get)(label)
# label is now tuple of 3 2d arrays 
# transpose to 3d array
unc_label = np.moveaxis(unc_label, 0, -1)
label = np.moveaxis(label, 0, -1)


fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].imshow(crop)
axes[0].imshow(unc_label, alpha=0.5, cmap="jet")
axes[0].set_title("Before SR")
axes[0].axis("off")
axes[1].imshow(crop)
axes[1].imshow(label, alpha=0.5, cmap="jet")
axes[1].set_title("After SR")
axes[1].axis("off")
# legend

plt.legend(handles=patches, bbox_to_anchor=(1.05, 0.5),
           loc='center left', fontsize=8,
           borderaxespad=0.)
plt.suptitle("Per-patch mask comparison (pre vs post SR)")
plt.tight_layout()
plt.show()
```

```python
"""Cityscapes semantic‑segmentation evaluation pipeline (enhanced).

All configurable parameters are collected in the CONFIG section below so they can be
adjusted without touching the rest of the code.  The script now *automatically*
visualises a sample image for every IoU calculation step (raw prediction, thresholded
prediction, and refined/patched prediction).  The sample shown can be selected via the
new ``SAMPLE_INDEX`` configuration flag.

Author: <your‑name>
Date: 2025‑05‑20
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"          # Cityscapes root directory
CHECKPOINT_PATH: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"  # Trained weights
CHECKPOINT_PATH_SPLICE: str = "./checkpoints_sr_splice/last.pth"  # Trained weights
OUTPUT_STRIDE: int = 16                               # DeeplabV3+ output stride
NUM_CLASSES: int = 19                                 # Number of train IDs in Cityscapes
SMALL_OBJECT_CLASSES: list[int] = [6, 7]              # Classes to super‑resolve/patch
PRED_CONF_THRESHOLD: float = 0.5                      # Probability threshold before argmax
MORPH_KERNEL_SIZE: int = 20                           # Kernel size for morphological closing
MIN_INSTANCE_SIDE: int = 50                           # Minimum w/h of cropped instance (px)
SR_MODEL_NAME: str = "caidas/swin2SR-lightweight-x2-64"  # HuggingFace model for SR
SAMPLE_INDEX: int = 1                                 # Index of the validation image to
                                                      # *visualise and evaluate*

# Choose device (CPU fallback)
import torch
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================
# LIBRARIES & MODEL LOADING
# ============================
from pathlib import Path
from typing import Iterable, Tuple, List

import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from torchvision import transforms as T
from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution

from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes  # dataset definitions used by the model
from torchvision.datasets import Cityscapes  # torchvision dataset for image loading

# ----------------------------
# Helper dicts derived from the dataset metadata
# ----------------------------
CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# ============================
# UTILITY FUNCTIONS
# ============================

def calculate_iou(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> np.ndarray:
    """Compute IoU per class (ignoring empty classes)."""
    ious: List[float] = []
    for cls in range(num_classes):
        intersection = np.logical_and(pred == cls, true == cls).sum()
        union = np.logical_or(pred == cls, true == cls).sum()
        if union > 0:
            ious.append(intersection / union)
    return np.asarray(ious, dtype=float)


def closing(mask: np.ndarray, kernel_size: int = MORPH_KERNEL_SIZE) -> np.ndarray:
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_CLOSE, kernel)


def crop_with_padding(image: np.ndarray, x: int, y: int, w: int, h: int, padding: float = 0.1) -> np.ndarray:
    """Crop an image with relative padding expressed as a fraction of (w, h)."""
    pad_x, pad_y = int(w * padding), int(h * padding)
    x1, y1 = max(0, x - pad_x), max(0, y - pad_y)
    x2, y2 = min(image.shape[1], x + w + pad_x), min(image.shape[0], y + h + pad_y)
    return image[y1:y2, x1:x2]

# ============================
# VISUALISATION HELPERS
# ============================

def show_image(image: np.ndarray, title: str | None = None):  # noqa: D401
    """Quick image viewer."""
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis("off")
    plt.show()


def show_image_with_mask(image: np.ndarray, mask: np.ndarray, opacity: float = 0.5, title: str | None = None):
    plt.imshow(image)
    im = plt.imshow(mask, alpha=opacity, cmap="jet")
    plt.axis("off")
    if title:
        plt.title(title)

    unique_vals = np.unique(mask)
    patches = [mpatches.Patch(color=im.cmap(im.norm(v)),
                              label=f"{CITYSCAPES_LABELS.get(int(v), str(int(v)))} ({int(v)})")
               for v in unique_vals]
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc="upper left", borderaxespad=0.)
    plt.show()

# ============================
# MAIN PIPELINE
# ============================

def main():
    # ---- Load network -----------------------------------------------------
    model = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES,
                                    output_stride=OUTPUT_STRIDE,
                                    pretrained_backbone=False)
    ckpt = torch.load(CHECKPOINT_PATH, map_location=DEVICE, weights_only=False)
    model.load_state_dict(ckpt["model_state"], strict=False)
    model.eval().to(DEVICE)

    # ---- Dataset sample ---------------------------------------------------
    dataset = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
    img_pil, mask_pil = dataset[SAMPLE_INDEX]  # PIL images

    # Convert to NumPy & label ids → train ids
    image_np = np.array(img_pil)
    mask_np = np.vectorize(ID2TRAINID.get)(np.array(mask_pil))

    # ---- Forward pass -----------------------------------------------------
    transform = T.Compose([
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    inp = transform(img_pil.convert("RGB")).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        logits = model(inp)
    logits = logits[0]
    logits = (logits - logits.min()) / (logits.max() - logits.min())  # [0, 1]
    pred_map = logits.cpu().numpy()

    # ----------------------------------------------------------------------
    # STEP 1 – Raw prediction (no thresholding)
    # ----------------------------------------------------------------------
    raw_pred_labels = pred_map.argmax(axis=0).astype(np.uint8)
    show_image_with_mask(image_np, raw_pred_labels, opacity=0.5, title="Raw prediction map")
    print(f"Mean IoU (raw): {calculate_iou(raw_pred_labels, mask_np).mean():.4f}")

    # ----------------------------------------------------------------------
    # STEP 2 – Confidence thresholding
    # ----------------------------------------------------------------------
    pred_map_thr = np.where(pred_map > PRED_CONF_THRESHOLD, pred_map, 0.0)
    pred_labels = pred_map_thr.argmax(axis=0).astype(np.uint8)
    show_image_with_mask(image_np, pred_labels, opacity=0.5,
                         title=f"Thresholded prediction (>{PRED_CONF_THRESHOLD})")
    print(f"Mean IoU (after threshold {PRED_CONF_THRESHOLD}): {calculate_iou(pred_labels, mask_np).mean():.4f}")

    # ----------------------------------------------------------------------
    # STEP 3 – Morphological closing (refinement)
    # ----------------------------------------------------------------------
    closed_labels = closing(pred_labels, kernel_size=MORPH_KERNEL_SIZE)
    print(f"Mean IoU (after closing): {calculate_iou(closed_labels, mask_np).mean():.4f}")

    # ----------------------------------------------------------------------
    # Extract small instances for SR + local re‑inference -------------------
    # ----------------------------------------------------------------------
    instances: list[tuple[int, np.ndarray, Tuple[int, int, int, int]]] = []
    for class_id in SMALL_OBJECT_CLASSES:
        class_mask = (closed_labels == class_id).astype(np.uint8)
        contours, _ = cv2.findContours(class_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE:
                continue  # skip small islands
            crop = crop_with_padding(image_np, x, y, w, h, padding=0.1)
            instances.append((class_id, crop, (x, y, w, h)))

    # ---- Super‑resolution model ------------------------------------------
    processor = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
    sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()

    def super_resolve(image: np.ndarray) -> np.ndarray:
        inputs = processor(image, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            outputs = sr_model(**inputs)
        sr = outputs.reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
        sr = np.moveaxis(sr, 0, -1)
        return (sr * 255).round().astype(np.uint8)

    # ---- SR + local re‑inference -----------------------------------------
    patched_labels = pred_labels.copy()
    for class_id, crop, (x, y, w, h) in instances:
        sr_crop = super_resolve(crop.transpose(2, 0, 1))
        tensor_crop = torch.from_numpy(sr_crop.transpose(2, 0, 1)).float().unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            logits_crop = model(tensor_crop)[0]
        probs_crop = torch.softmax(logits_crop, dim=0).cpu().numpy()
        new_labels = probs_crop.argmax(axis=0)
        new_labels = cv2.resize(new_labels, (w, h), interpolation=cv2.INTER_NEAREST)
        patched_labels[y:y + h, x:x + w] = new_labels

    # ----------------------------------------------------------------------
    # STEP 4 – Patched prediction (final refined map)
    # ----------------------------------------------------------------------
    show_image_with_mask(image_np, patched_labels, opacity=0.5, title="Patched prediction map")
    iou_final = calculate_iou(patched_labels, mask_np).mean()
    print(f"Mean IoU (patched): {iou_final:.4f}")


if __name__ == "__main__":
    main()
```

```python
"""Cityscapes semantic-segmentation evaluation pipeline (batched).

Changes from the previous version
---------------------------------
* **Full validation sweep** – iterates through the entire *val* split with a configurable
  batch size.
* **tqdm progress bar** – shows real-time progress over batches.
* **Metrics list** – per-image IoU statistics are appended to `metrics_log` so they can
  be inspected or saved after the run.
* `torch.load(..., weights_only=False)` – keeps parity with the original training code.

All visualisation calls remain commented-out so the script runs headless by default.
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"                                # Cityscapes root
CHECKPOINT_PATH: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"  # Trained weights
OUTPUT_STRIDE: int = 16                                                     # DeeplabV3+ output stride
NUM_CLASSES: int = 19                                                       # Number of train IDs
BATCH_SIZE: int = 4                                                         # Images per forward pass
SMALL_OBJECT_CLASSES: list[int] = [6, 7]                                    # Classes to refine
PRED_CONF_THRESHOLD: float = 0.5                                            # Confidence cut-off
MORPH_KERNEL_SIZE: int = 20                                                 # Closing kernel size
MIN_INSTANCE_SIDE: int = 50                                                 # Min w/h of crops (px)
SR_MODEL_NAME: str = "caidas/swin2SR-lightweight-x2-64"                      # HF SR model id
NUM_WORKERS: int = 4                                                        # DataLoader workers
SEED: int = 42                                                              # Reproducibility

# Choose device (CPU fallback)
import torch
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================
# LIBRARIES & MODEL LOADING
# ============================
from pathlib import Path
from typing import List, Dict, Tuple, Any

import numpy as np
import cv2
from tqdm.auto import tqdm
import matplotlib.pyplot as plt  # noqa: F401  – only used if you uncomment visuals
import matplotlib.patches as mpatches  # noqa: F401
from torchvision import transforms as T
from torch.utils.data import DataLoader
from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution

from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes

# ----------------------------
# Helper dicts derived from dataset metadata
# ----------------------------
CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# Set RNG seed for deterministic DataLoader order
import random
import os
random.seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
np.random.seed(SEED)

torch.manual_seed(SEED)
if DEVICE.type == "cuda":
    torch.cuda.manual_seed_all(SEED)

# ============================
# UTILITY FUNCTIONS
# ============================

def calculate_iou(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> np.ndarray:
    """Compute IoU per class (ignore empty classes)."""
    ious: List[float] = []
    for cls in range(num_classes):
        intersection = np.logical_and(pred == cls, true == cls).sum()
        union = np.logical_or(pred == cls, true == cls).sum()
        if union > 0:
            ious.append(intersection / union)
    return np.asarray(ious, dtype=float)


def closing(mask: np.ndarray, kernel_size: int = MORPH_KERNEL_SIZE) -> np.ndarray:
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_CLOSE, kernel)


def crop_with_padding(image: np.ndarray, x: int, y: int, w: int, h: int, padding: float = 0.1) -> np.ndarray:
    """Crop an area with additional relative padding."""
    pad_x, pad_y = int(w * padding), int(h * padding)
    x1, y1 = max(0, x - pad_x), max(0, y - pad_y)
    x2, y2 = min(image.shape[1], x + w + pad_x), min(image.shape[0], y + h + pad_y)
    return image[y1:y2, x1:x2]


# ============================
# VISUALISATION HELPERS (OPTIONAL)
# ============================

def show_image(image: np.ndarray, title: str | None = None):  # noqa: D401
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis("off")
    plt.show()


def show_image_with_mask(image: np.ndarray, mask: np.ndarray, opacity: float = 0.5, title: str | None = None):  # noqa: D401
    plt.imshow(image)
    overlay = plt.imshow(mask, alpha=opacity, cmap="jet")
    plt.axis("off")
    if title:
        plt.title(title)
    unique_vals = np.unique(mask)
    patches = [mpatches.Patch(color=overlay.cmap(overlay.norm(v)), label=f"{CITYSCAPES_LABELS.get(int(v), v)} ({int(v)})") for v in unique_vals]
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.show()


# ============================
# MAIN PIPELINE
# ============================

def main() -> None:
    # ---- Model -----------------------------------------------------------
    model = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES,
                                    output_stride=OUTPUT_STRIDE,
                                    pretrained_backbone=False)
    ckpt = torch.load(CHECKPOINT_PATH, map_location=DEVICE, weights_only=False)
    model.load_state_dict(ckpt["model_state"], strict=False)
    model.eval().to(DEVICE)

    # ---- Super-resolution model -----------------------------------------
    processor = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
    sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()

    def super_resolve(image: np.ndarray) -> np.ndarray:
        inputs = processor(image, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            outputs = sr_model(**inputs)
        sr = outputs.reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
        return (np.moveaxis(sr, 0, -1) * 255).round().astype(np.uint8)

    # ---- Dataset & DataLoader ------------------------------------------
    dataset = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")

    def collate(batch):
        imgs, masks = zip(*batch)
        return list(imgs), list(masks)

    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=collate)

    # ---- Preprocessing transform ---------------------------------------
    transform = T.Compose([
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])

    # ---- Metrics storage ------------------------------------------------
    metrics_log: List[Dict[str, Any]] = []
    global_idx = 0

    # ---- Iterate over batches ------------------------------------------
    for imgs_pil, masks_pil in tqdm(loader, desc="Inference", unit="batch"):
        # Convert PIL to tensors for batch inference
        batch_tensor = torch.stack([transform(pil.convert("RGB")) for pil in imgs_pil]).to(DEVICE)
        with torch.no_grad():
            logits_batch = model(batch_tensor)

        for b in range(len(imgs_pil)):
            img_np = np.array(imgs_pil[b])
            mask_np = np.vectorize(ID2TRAINID.get)(np.array(masks_pil[b]))
            logits = logits_batch[b]
            logits = (logits - logits.min()) / (logits.max() - logits.min())  # [0,1]
            logits_np = logits.cpu().numpy()

            # ---------------- Raw prediction ----------------
            raw_pred = logits_np.argmax(axis=0).astype(np.uint8)
            iou_raw = calculate_iou(raw_pred, mask_np).mean()

            # ---------------- Thresholding ------------------
            thresh_pred = np.where(logits_np > PRED_CONF_THRESHOLD, logits_np, 0.0).argmax(axis=0).astype(np.uint8)
            iou_thresh = calculate_iou(thresh_pred, mask_np).mean()

            # ---------------- Morph closing -----------------
            closed_pred = closing(thresh_pred)
            iou_closed = calculate_iou(closed_pred, mask_np).mean()

            # ---------------- SR + patch --------------------
            patched_pred = closed_pred.copy()
            # collect instances for selected classes
            instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]] = []
            for cls_id in SMALL_OBJECT_CLASSES:
                cls_mask = (closed_pred == cls_id).astype(np.uint8)
                contours, _ = cv2.findContours(cls_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                for cnt in contours:
                    x, y, w, h = cv2.boundingRect(cnt)
                    if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE:
                        continue
                    crop = crop_with_padding(img_np, x, y, w, h, padding=0.1)
                    instances.append((cls_id, crop, (x, y, w, h)))

            for cls_id, crop, (x, y, w, h) in instances:
                sr_crop = super_resolve(crop.transpose(2, 0, 1))
                tensor_crop = torch.from_numpy(sr_crop.transpose(2, 0, 1)).float().unsqueeze(0).to(DEVICE)
                with torch.no_grad():
                    logits_crop = model(tensor_crop)[0]
                probs_crop = torch.softmax(logits_crop, dim=0).cpu().numpy()
                new_labels = probs_crop.argmax(axis=0)
                new_labels = cv2.resize(new_labels, (w, h), interpolation=cv2.INTER_NEAREST)
                patched_pred[y:y + h, x:x + w] = new_labels

            iou_patched = calculate_iou(patched_pred, mask_np).mean()

            # ---------------- Log metrics -------------------
            metrics_log.append({
                "idx": global_idx,
                "iou_raw": float(iou_raw),
                "iou_thresh": float(iou_thresh),
                "iou_closed": float(iou_closed),
                "iou_patched": float(iou_patched),
            })
            global_idx += 1

    # ---- Print global summary ------------------------------------------
    def mean(k: str):
        return np.mean([m[k] for m in metrics_log])

    print("\n=========== Summary on val split ===========")
    print(f"Mean IoU raw      : {mean('iou_raw'):.4f}")
    print(f"Mean IoU threshold: {mean('iou_thresh'):.4f}")
    print(f"Mean IoU closed   : {mean('iou_closed'):.4f}")
    print(f"Mean IoU patched  : {mean('iou_patched'):.4f}")

    # metrics_log is now in memory – you can pickle, JSON-dump, or export it as needed.


if __name__ == "__main__":
    main()
```

```python
"""Cityscapes semantic‑segmentation evaluation pipeline – **batched** with optional
super‑resolution back‑ends (Swin2SR *or* Real‑ESRGAN).

Key features
------------
* Full validation‑set sweep with tqdm progress bar.
* Per‑image IoU metrics logged to ``metrics_log`` for post‑processing.
* Switchable SR back‑end via ``SR_BACKEND`` config variable:
  - ``'swin2sr'`` (default, HuggingFace *swin2SR-lightweight-x2-64*).
  - ``'realesrgan'`` (requires the *realesrgan* package; model weights are
    auto‑downloaded on first run).
* All other behaviour unchanged from previous iteration.
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"                                # Cityscapes root
CHECKPOINT_PATH: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"  # Trained weights
OUTPUT_STRIDE: int = 16                                                     # DeeplabV3+ output stride
NUM_CLASSES: int = 19                                                       # Number of train IDs
BATCH_SIZE: int = 4                                                         # Images per batch
SMALL_OBJECT_CLASSES: list[int] = [6, 7]                                    # Classes to refine
PRED_CONF_THRESHOLD: float = 0.5                                            # Confidence cut‑off
MORPH_KERNEL_SIZE: int = 20                                                 # Closing kernel size
MIN_INSTANCE_SIDE: int = 50                                                 # Min w/h for crops (px)

# ---- Super‑resolution ----------------------------------------------------
SR_BACKEND: str = "realesrgan"        # 'swin2sr' | 'realesrgan'
# Swin2SR (HuggingFace) params
SR_MODEL_NAME: str = "caidas/swin2SR-lightweight-x2-64"
# Real‑ESRGAN params (only used when SR_BACKEND == 'realesrgan')
SR_REALESRGAN_MODEL: str = "RealESRGAN_x2plus"  # see README for options
SR_REALESRGAN_DENOISE: float = 0.5
SR_REALESRGAN_TILE: int = 0
SR_REALESRGAN_TILE_PAD: int = 10
SR_REALESRGAN_OUTSCALE: int = 4
SR_REALESRGAN_HALF: bool = True  # use fp16 when CUDA available

NUM_WORKERS: int = 4             # DataLoader workers
SEED: int = 42                   # Reproducibility

# ============================
# ENV & DEPENDENCIES
# ============================
import os
import random
from pathlib import Path
from typing import List, Dict, Tuple, Any

import cv2
import numpy as np
import torch
from torch.utils.data import DataLoader
from torchvision import transforms as T
from tqdm.auto import tqdm

import matplotlib.pyplot as plt  # visuals remain commented‑out by default  noqa: F401
import matplotlib.patches as mpatches  # noqa: F401

from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes

# ---------- SR back‑end conditional imports ----------
if SR_BACKEND == "swin2sr":
    from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
elif SR_BACKEND == "realesrgan":
    from realesrgan import RealESRGANer
    from basicsr.archs.rrdbnet_arch import RRDBNet
    from basicsr.archs.srvgg_arch import SRVGGNetCompact
    from basicsr.utils.download_util import load_file_from_url
else:
    raise ValueError(f"Unsupported SR_BACKEND: {SR_BACKEND}")

# ---------- Device & RNG ----------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
random.seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if DEVICE.type == "cuda":
    torch.cuda.manual_seed_all(SEED)

# ============================
# DATASET META‑MAPPINGS
# ============================
CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# ============================
# UTILITIES
# ============================

def calculate_iou(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> np.ndarray:
    ious: List[float] = []
    for cls in range(num_classes):
        inter = np.logical_and(pred == cls, true == cls).sum()
        union = np.logical_or(pred == cls, true == cls).sum()
        if union > 0:
            ious.append(inter / union)
    return np.asarray(ious, dtype=float)


def closing(mask: np.ndarray, kernel_size: int = MORPH_KERNEL_SIZE) -> np.ndarray:
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_CLOSE, kernel)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    pad_x, pad_y = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - pad_x), max(0, y - pad_y)
    x2, y2 = min(img.shape[1], x + w + pad_x), min(img.shape[0], y + h + pad_y)
    return img[y1:y2, x1:x2]

# ============================
# VISUALISATION HELPERS (optional)
# ============================

def show_image(image: np.ndarray, title: str | None = None):  # noqa: D401
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis("off")
    plt.show()


def show_image_with_mask(image: np.ndarray, mask: np.ndarray, opacity: float = 0.5, title: str | None = None):  # noqa: D401
    plt.imshow(image)
    overlay = plt.imshow(mask, alpha=opacity, cmap="jet")
    if title:
        plt.title(title)
    plt.axis("off")
    # legend
    patches = [mpatches.Patch(color=overlay.cmap(overlay.norm(v)), label=f"{CITYSCAPES_LABELS.get(int(v), v)} ({int(v)})") for v in np.unique(mask)]
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.show()

# ============================
# SUPER‑RESOLUTION FACTORY
# ============================

def build_super_resolver():
    """Return a ``super_resolve`` callable that upsamples an RGB HWC *np.uint8* image."""

    if SR_BACKEND == "swin2sr":
        processor = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
        sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()

        def _swin2sr(img_rgb_hwc: np.ndarray) -> np.ndarray:
            inputs = processor(img_rgb_hwc, return_tensors="pt").to(DEVICE)
            with torch.no_grad():
                out = sr_model(**inputs).reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
            return (np.moveaxis(out, 0, -1) * 255).round().astype(np.uint8)

        return _swin2sr

    # ---------- Real‑ESRGAN path ----------
    # Map model name to arch class, netscale and download URL(s)
    def _realesrgan_model(model_name: str):
        if model_name == "RealESRGAN_x4plus":
            return RRDBNet(3, 3, 64, 23, 32, scale=4), 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth"]
        if model_name == "RealESRNet_x4plus":
            return RRDBNet(3, 3, 64, 23, 32, scale=4), 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.1/RealESRNet_x4plus.pth"]
        if model_name == "RealESRGAN_x2plus":
            return RRDBNet(3, 3, 2, 64, 23, 32), 2, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.1/RealESRGAN_x2plus.pth"]
        if model_name == "realesr-general-x4v3":
            model = SRVGGNetCompact(3, 3, 64, num_conv=32, upscale=4, act_type="prelu")
            # weight pairs for dni blending handled later
            return model, 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-general-wdn-x4v3.pth",
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-general-x4v3.pth",
            ]
        # fallback – anime, etc. (XS size)
        return SRVGGNetCompact(3, 3, 64, num_conv=16, upscale=4, act_type="prelu"), 4, [
            "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-animevideov3.pth"]

    model_arch, netscale, urls = _realesrgan_model(SR_REALESRGAN_MODEL)
    weights_dir = "weights"
    os.makedirs(weights_dir, exist_ok=True)
    model_paths = []
    for url in urls:
        fname = url.split("/")[-1]
        local_p = os.path.join(weights_dir, fname)
        if not os.path.exists(local_p):
            local_p = load_file_from_url(url, model_dir=str(weights_dir), progress=True)
        model_paths.append(str(local_p))

    dni_weight = None
    if SR_REALESRGAN_MODEL == "realesr-general-x4v3" and SR_REALESRGAN_DENOISE != 1:
        # Expect first path is clean, second is wdn
        dni_weight = [SR_REALESRGAN_DENOISE, 1 - SR_REALESRGAN_DENOISE]

    upsampler = RealESRGANer(
        scale=netscale,
        model_path=model_paths if len(model_paths) > 1 else model_paths[0],
        dni_weight=dni_weight,
        model=model_arch,
        tile=SR_REALESRGAN_TILE,
        tile_pad=SR_REALESRGAN_TILE_PAD,
        half=SR_REALESRGAN_HALF and DEVICE.type == "cuda",
        gpu_id=0 if DEVICE.type == "cuda" else None,
    )

    def _realesrgan(img_rgb_hwc: np.ndarray) -> np.ndarray:
        img_bgr = cv2.cvtColor(img_rgb_hwc, cv2.COLOR_RGB2BGR)
        output_bgr, _ = upsampler.enhance(img_bgr, outscale=SR_REALESRGAN_OUTSCALE)
        return cv2.cvtColor(output_bgr, cv2.COLOR_BGR2RGB)

    return _realesrgan

# ============================
# MAIN PIPELINE
# ============================

def main() -> None:
    # ---- Network --------------------------------------------------------
    net = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES, output_stride=OUTPUT_STRIDE, pretrained_backbone=False)
    ckpt = torch.load(CHECKPOINT_PATH, map_location=DEVICE, weights_only=False)
    net.load_state_dict(ckpt["model_state"], strict=False)
    net.eval().to(DEVICE)

    # ---- Super‑res back‑end --------------------------------------------
    super_resolve = build_super_resolver()

    # ---- Dataset & DataLoader ------------------------------------------
    dataset = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")

    def collate(batch):
        imgs, masks = zip(*batch)
        return list(imgs), list(masks)

    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=collate)

    preprocess = T.Compose([
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])

    metrics_log: List[Dict[str, Any]] = []
    global_idx = 0

    for imgs_pil, masks_pil in tqdm(loader, desc="Inference", unit="batch"):
        inp = torch.stack([preprocess(p.convert("RGB")) for p in imgs_pil]).to(DEVICE)
        with torch.no_grad():
            logits_batch = net(inp)

        for b in range(len(imgs_pil)):
            img_np = np.array(imgs_pil[b])  # HWC, RGB
            mask_np = np.vectorize(ID2TRAINID.get)(np.array(masks_pil[b]))
            logits = logits_batch[b]
            logits = (logits - logits.min()) / (logits.max() - logits.min())
            logits_np = logits.cpu().numpy()

            raw_pred = logits_np.argmax(axis=0).astype(np.uint8)
            iou_raw = calculate_iou(raw_pred, mask_np).mean()

            thresh_pred = np.where(logits_np > PRED_CONF_THRESHOLD, logits_np, 0.)
            thresh_pred = thresh_pred.argmax(axis=0).astype(np.uint8)
            iou_thresh = calculate_iou(thresh_pred, mask_np).mean()

            closed_pred = closing(thresh_pred)
            iou_closed = calculate_iou(closed_pred, mask_np).mean()

            patched_pred = closed_pred.copy()
            # --- small‑object refinement ---
            instances: List[Tuple[int, np.ndarray, Tuple[int, int, int, int]]] = []
            for cid in SMALL_OBJECT_CLASSES:
                mask_cls = (closed_pred == cid).astype(np.uint8)
                contours, _ = cv2.findContours(mask_cls, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                for cnt in contours:
                    x, y, w, h = cv2.boundingRect(cnt)
                    if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE:
                        continue
                    crop = crop_with_padding(img_np, x, y, w, h)
                    instances.append((cid, crop, (x, y, w, h)))

            for cid, crop_rgb, (x, y, w, h) in instances:
                sr_crop = super_resolve(crop_rgb)  # HWC RGB uint8
                tensor_crop = torch.from_numpy(sr_crop.transpose(2, 0, 1)).float().unsqueeze(0).to(DEVICE)
                with torch.no_grad():
                    crop_logits = net(tensor_crop)[0]
                probs = torch.softmax(crop_logits, dim=0).cpu().numpy()
                new_lbls = probs.argmax(axis=0)
                new_lbls = cv2.resize(new_lbls, (w, h), interpolation=cv2.INTER_NEAREST)
                patched_pred[y:y + h, x:x + w] = new_lbls

            iou_patched = calculate_iou(patched_pred, mask_np).mean()

            metrics_log.append({
                "idx": global_idx,
                "iou_raw": float(iou_raw),
                "iou_thresh": float(iou_thresh),
                "iou_closed": float(iou_closed),
                "iou_patched": float(iou_patched),
            })
            global_idx += 1

    # ---- Summary ---------------------------------------------------------
    def mean(key: str):
        return np.mean([m[key] for m in metrics_log])

    print("\n=========== Validation Summary ===========")
    print(f"Mean IoU raw      : {mean('iou_raw'):.4f}")
    print(f"Mean IoU threshold: {mean('iou_thresh'):.4f}")
    print(f"Mean IoU closed   : {mean('iou_closed'):.4f}")
    print(f"Mean IoU patched  : {mean('iou_patched'):.4f}")

    # metrics_log ready for external use (e.g., save to JSON/CSV)


if __name__ == "__main__":
    main()
```

```python
"""
Cityscapes semantic-segmentation evaluation
==========================================
Runs **five presets** sequentially (requested order):

1. **DeepLabV3+** + *Swin2SR Classical ×2*
2. **DeepLabV3+** + *Swin2SR Lightweight ×2*
3. **SegFormer-B1** + *RealESRGAN ×4*
4. **SegFormer-B1** + *RealESRGAN ×2*
5. **SegFormer-B1** + *Swin2SR Lightweight ×2*

For each preset the script performs two-sided confidence screening, morphological
closing→opening, optional super-resolution (one-by-one for Swin2SR to avoid
padding), per-class IoU logging (including **`iou_raw`**), and writes a
self-describing JSON.
"""

# ============= PRESETS ===================================================
PRESETS: list[dict] = [
    # 1 ▸ DeepLabV3+ + Swin2SR Classical ×2
    {"backbone":"deeplab",  "sr_backend":"swin2sr",   "sr_model":"caidas/swin2SR-classical-sr-x2-64"},

    # 2 ▸ DeepLabV3+ + Swin2SR Lightweight ×2
    {"backbone":"deeplab",  "sr_backend":"swin2sr",   "sr_model":"caidas/swin2SR-lightweight-x2-64"},

    # 3 ▸ SegFormer-B1 + RealESRGAN ×4
    {"backbone":"segformer","sr_backend":"realesrgan","sr_model":"RealESRGAN_x4plus","sr_scale":4},

    # 4 ▸ SegFormer-B1 + RealESRGAN ×2
    {"backbone":"segformer","sr_backend":"realesrgan","sr_model":"RealESRGAN_x2plus","sr_scale":2},

    # 5 ▸ SegFormer-B1 + Swin2SR Lightweight ×2
    {"backbone":"segformer","sr_backend":"swin2sr",   "sr_model":"caidas/swin2SR-lightweight-x2-64"},
]
# ========================================================================

# ---------- COMMON HYPERPARAMS ------------------------------------------
DATASET_ROOT            = "./datasets/cityscapes"
DEEPLAB_CHECKPOINT      = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"
SEGFORMER_MODEL_NAME    = "nvidia/segformer-b1-finetuned-cityscapes-1024-1024"
NUM_CLASSES             = 19
BATCH_SIZE              = 4
OUTPUT_STRIDE           = 16          # DeepLab only
SMALL_OBJECT_CLASSES    = [6, 7]
HIGH_CONF_THRESH        = 0.5
LOW_CONF_THRESH         = 0.2
MORPH_KERNEL_SIZE       = 20
MIN_INSTANCE_SIDE       = 50
NUM_WORKERS             = 4
SEED                    = 42

import os, random, json, datetime
from pathlib import Path
from typing import List, Dict, Any, Optional
import cv2, numpy as np, torch
from torch.utils.data import DataLoader
from torchvision import transforms as T
from tqdm.auto import tqdm
from PIL import Image

from deeplabv3plus.datasets import Cityscapes as DeeplabCity
from torchvision.datasets import Cityscapes
CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCity.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCity.classes}
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
if DEVICE.type=="cuda": torch.cuda.manual_seed_all(SEED)

# ---------------- helper fns --------------------------------------------

iou_pc=lambda p,t,n=NUM_CLASSES:[((i:=np.logical_and(p==c,t==c).sum()),(u:=np.logical_or(p==c,t==c).sum()))and(i/u)if u else None for c in range(n)]
morph=lambda m,op: cv2.morphologyEx(m.astype(np.uint8),op,np.ones((MORPH_KERNEL_SIZE,)*2,np.uint8))
closing=lambda m:morph(m,cv2.MORPH_CLOSE); opening=lambda m:morph(m,cv2.MORPH_OPEN)
cpad=lambda im,x,y,w,h,p=0.1:(im[max(0,y-int(h*p)):min(im.shape[0],y+h+int(h*p)),max(0,x-int(w*p)):min(im.shape[1],x+w+int(w*p))])

# ---------------- backbone factories ------------------------------------

def build_deeplab():
    from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
    net=deeplabv3plus_mobilenet(num_classes=NUM_CLASSES,output_stride=OUTPUT_STRIDE,pretrained_backbone=False)
    net.load_state_dict(torch.load(DEEPLAB_CHECKPOINT,map_location=DEVICE,weights_only=False)["model_state"],strict=False)
    net.to(DEVICE).eval()
    tfm=T.Compose([T.ToTensor(),T.Normalize(mean=[0.485,0.456,0.406],std=[0.229,0.224,0.225])])
    return net, (lambda p: tfm(p.convert("RGB"))), (lambda bt: net(bt))

def build_segformer():
    from transformers import SegformerFeatureExtractor, SegformerForSemanticSegmentation
    ext=SegformerFeatureExtractor.from_pretrained(SEGFORMER_MODEL_NAME)
    net=SegformerForSemanticSegmentation.from_pretrained(SEGFORMER_MODEL_NAME).to(DEVICE).eval()
    def _fwd(pil_lst):
        inp=ext(images=pil_lst,return_tensors="pt",padding=True).to(DEVICE)
        with torch.no_grad(): out=net(**inp).logits
        _,_,h4,w4=out.shape
        return torch.nn.functional.interpolate(out,size=(h4*4,w4*4),mode="bilinear",align_corners=False)
    return net,ext,_fwd

# ---------------- SR factories -----------------------------------------

def swin2sr_single(model_name:str):
    from transformers import AutoImageProcessor,Swin2SRForImageSuperResolution
    proc=AutoImageProcessor.from_pretrained(model_name); net=Swin2SRForImageSuperResolution.from_pretrained(model_name).to(DEVICE).eval()
    return lambda img:(lambda rec:(np.moveaxis(rec.squeeze(),0,-1)*255).round().astype(np.uint8))((lambda t: net(**t).reconstruction.clamp_(0,1).detach().cpu().numpy())(proc(images=img,return_tensors="pt").to(DEVICE)))

def realesr_single(model_name:str,scale:int):
    from realesrgan import RealESRGANer
    from basicsr.archs.rrdbnet_arch import RRDBNet
    from basicsr.utils.download_util import load_file_from_url
    arch,url=(RRDBNet(3,3,4 if scale==4 else 2,64,23,32),{"RealESRGAN_x4plus":"https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth","RealESRGAN_x2plus":"https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.1/RealESRGAN_x2plus.pth"}[model_name])
    w=load_file_from_url(url,model_dir="weights",progress=True)
    up=RealESRGANer(scale=scale,model_path=w,model=arch,gpu_id=0 if DEVICE.type=="cuda" else None,half=False)
    return lambda img: cv2.cvtColor(up.enhance(cv2.cvtColor(img,cv2.COLOR_RGB2BGR),outscale=scale)[0],cv2.COLOR_BGR2RGB)

# ---------------- run one preset ---------------------------------------

def run(cfg:dict):
    now=datetime.datetime.now().strftime("%Y%m%d_%H%M%S"); tag=f"{cfg['backbone']}_{cfg['sr_backend']}_{now}"; fname=f"metrics_{tag}.json"
    # backbone
    if cfg['backbone']=='deeplab': net,prep,fwd=build_deeplab()
    else: net,ext,fwd=build_segformer()
    # SR
    if cfg['sr_backend']=='swin2sr': sr=swin2sr_single(cfg['sr_model'])
    elif cfg['sr_backend']=='realesrgan': sr=realesr_single(cfg['sr_model'],cfg['sr_scale'])
    else: sr=lambda im:im
    ds=Cityscapes(DATASET_ROOT,split='val',mode='fine',target_type='semantic')
    loader=DataLoader(ds,batch_size=BATCH_SIZE,shuffle=False,num_workers=NUM_WORKERS,collate_fn=lambda b:tuple(zip(*b)))
    metrics=[]; gid=0
    for imgs,masks in tqdm(loader,desc=tag,unit='batch'):
        logits_b=fwd(list(imgs)) if cfg['backbone']=='segformer' else fwd(torch.stack([prep(p) for p in imgs]).to(DEVICE))
        for i in range(len(imgs)):
            img=np.array(imgs[i]); gt=np.vectorize(ID2TRAINID.get)(np.array(masks[i])); logits=logits_b[i]
            if logits.shape[1:]!=img.shape[:2]: logits=torch.nn.functional.interpolate(logits.unsqueeze(0),size=img.shape[:2],mode='bilinear',align_corners=False)[0]
            lnp=logits.detach().cpu().numpy(); raw=lnp.argmax(0).astype(np.uint8)
            maxp=lnp.max(0); high=(lnp*(maxp>=HIGH_CONF_THRESH)).argmax(0).astype(np.uint8); low=(lnp*(maxp<=LOW_CONF_THRESH)).argmax(0).astype(np.uint8)
            closed=closing(high); opened=opening(closed); refined=opened.copy()
            for cid in SMALL_OBJECT_CLASSES:
                cnts,_=cv2.findContours((opened==cid).astype(np.uint8),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
                for c in cnts:
                    x,y,w,h=cv2.boundingRect(c)
                    if w<MIN_INSTANCE_SIDE or h<MIN_INSTANCE_SIDE: continue
                    crop=cpad(img,x,y,w,h); sr_crop=sr(crop)
                    if cfg['backbone']=='segformer':
                        inp=ext(images=Image.fromarray(sr_crop),return_tensors='pt',padding=True).to(DEVICE); cl=net(**inp).logits
                        cl=torch.nn.functional.interpolate(cl,size=sr_crop.shape[:2],mode='bilinear',align_corners=False)[0]
                    else:
                        ct=T.Normalize(mean=[0.485,0.456,0.406],std=[0.229,0.224,0.225])(T.ToTensor()(sr_crop)).unsqueeze(0).to(DEVICE); cl=net(ct)[0]
                    lbl=torch.softmax(cl,0).detach().cpu().numpy().argmax(0); refined[y:y+h,x:x+w]=cv2.resize(lbl,(w,h),interpolation=cv2.INTER_NEAREST)
            metrics.append({"idx":gid,"iou_raw":iou_pc(raw,gt),"iou_high":iou_pc(high,gt),"iou_low":iou_pc(low,gt),"iou_closed":iou_pc(closed,gt),"iou_opened":iou_pc(opened,gt),"iou_refined":iou_pc(refined,gt)}); gid+=1
    with open(fname,'w',encoding='utf-8') as f: json.dump({"run_params":cfg,"metrics":metrics},f,indent=2)
    print('saved',fname)

# ----------------------------- main --------------------------------------
if __name__=='__main__':
    for p in PRESETS: run(p)
```

```python
"""Cityscapes semantic‑segmentation evaluation – batched, multi‑backbone (DeepLabV3+
| SegFormer), optional super‑resolution, *reverse+forward* confidence screening,
*morphological closing → opening* cleanup, per‑class metrics, self‑describing
JSON output.

What’s new (May 2025 rev)
-------------------------
* **Two‑sided confidence thresholding**
  * `HIGH_CONF_THRESH` keeps strong predictions (as before).
  * `LOW_CONF_THRESH` isolates *low*‑confidence pixels ("reverse threshold").
  * Both maps are logged for analysis.
* **Morphological opening after closing** to remove small islands once holes
  are filled.
* Metrics now include IoU for the *opened* mask and the isolated low‑conf mask.
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"

# ---- Backbone -----------------------------------------------------------
CHECKPOINT_PATH_SPLICE: str = "./checkpoints_sr_splice/last.pth"  # Trained weights
MODEL_BACKBONE: str = "deeplab"          # 'deeplab' | 'segformer'
DEEPLAB_CHECKPOINT: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"
SEGFORMER_MODEL_NAME: str = "nvidia/segformer-b1-finetuned-cityscapes-1024-1024"
OUTPUT_STRIDE: int = 16                  # only for DeepLab

NUM_CLASSES: int = 19
BATCH_SIZE: int = 4
SMALL_OBJECT_CLASSES: list[int] = [6, 7]
HIGH_CONF_THRESH: float = 0.5            # forward threshold
LOW_CONF_THRESH: float = 0.4             # reverse threshold
MORPH_KERNEL_SIZE: int = 20              # for closing & opening
MIN_INSTANCE_SIDE: int = 50

# ---- Super‑resolution ----------------------------------------------------
SR_BACKEND: str | None = "swin2sr"        # 'swin2sr' | 'realesrgan' | None
SR_MODEL_NAME: str = "caidas/swin2SR-lightweight-x2-64"
SR_REALESRGAN_MODEL: str = "RealESRGAN_x4plus"
SR_REALESRGAN_DENOISE: float = 0.5
SR_REALESRGAN_TILE: int = 0
SR_REALESRGAN_TILE_PAD: int = 10
SR_REALESRGAN_OUTSCALE: int = 4
SR_REALESRGAN_HALF: bool = True

# ---- Runtime ------------------------------------------------------------
NUM_WORKERS: int = 4
SEED: int = 42

# ============================
# ENV & DEPENDENCIES
# ============================
import os, random, json, datetime
from pathlib import Path
from typing import List, Dict, Tuple, Any, Optional

import cv2, numpy as np, torch
from torch.utils.data import DataLoader
from torchvision import transforms as T
from tqdm.auto import tqdm
from PIL import Image

# visuals (comment‑out callers by default)
import matplotlib.pyplot as plt  # noqa: F401
import matplotlib.patches as mpatches  # noqa: F401

# Backbone‑specific ---------------------------------------
if MODEL_BACKBONE == "deeplab":
    from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
elif MODEL_BACKBONE == "segformer":
    from transformers import SegformerFeatureExtractor, SegformerForSemanticSegmentation
else:
    raise ValueError("Unsupported MODEL_BACKBONE")

# Super‑res imports --------------------------------------
if SR_BACKEND == "swin2sr":
    from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
elif SR_BACKEND == "realesrgan":
    from realesrgan import RealESRGANer
    from basicsr.archs.rrdbnet_arch import RRDBNet
    from basicsr.archs.srvgg_arch import SRVGGNetCompact
    from basicsr.utils.download_util import load_file_from_url

# Dataset meta -------------------------------------------
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes

CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# Device & RNG -------------------------------------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
if DEVICE.type == "cuda":
    torch.cuda.manual_seed_all(SEED)

# ============================
# UTILITIES
# ============================

def calculate_iou_all(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> List[Optional[float]]:
    ious: List[Optional[float]] = []
    for c in range(num_classes):
        inter = np.logical_and(pred == c, true == c).sum()
        union = np.logical_or(pred == c, true == c).sum()
        ious.append(inter / union if union else None)
    return ious


def morph(mask: np.ndarray, op: int, ksize: int = MORPH_KERNEL_SIZE):
    kernel = np.ones((ksize, ksize), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), op, kernel)

closing = lambda m: morph(m, cv2.MORPH_CLOSE)
opening = lambda m: morph(m, cv2.MORPH_OPEN)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    px, py = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - px), max(0, y - py)
    x2, y2 = min(img.shape[1], x + w + px), min(img.shape[0], y + h + py)
    return img[y1:y2, x1:x2]

# ============================
# SUPER‑RESOLUTION FACTORY
# ============================

def build_super_resolver():
    if SR_BACKEND is None:
        return lambda img: img

    if SR_BACKEND == "swin2sr":
        proc = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
        sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()
        def _swin(img):
            t = proc(img, return_tensors="pt").to(DEVICE)
            with torch.no_grad():
                out = sr_model(**t).reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
            return (np.moveaxis(out, 0, -1) * 255).round().astype(np.uint8)
        return _swin

    # Real‑ESRGAN
    def _realesr_arch(name):
        if name == "RealESRGAN_x4plus":
            return RRDBNet(3, 3, 4, 64, 23, 32), 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth"]
        if name == "RealESRGAN_x2plus":
            return RRDBNet(3, 3, 2, 64, 23, 32), 2, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.1/RealESRGAN_x2plus.pth"]
        return SRVGGNetCompact(3, 3, 64, num_conv=32, upscale=4, act_type="prelu"), 4, [
            "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-general-x4v3.pth"]

    arch, netscale, urls = _realesr_arch(SR_REALESRGAN_MODEL)
    wdir = Path("weights"); wdir.mkdir(exist_ok=True)
    model_paths = [str(load_file_from_url(u, model_dir=str(wdir), progress=True)) for u in urls]
    upsampler = RealESRGANer(scale=netscale, model_path=model_paths[0] if len(model_paths) == 1 else model_paths,
                             model=arch, tile=SR_REALESRGAN_TILE, tile_pad=SR_REALESRGAN_TILE_PAD,
                             half=SR_REALESRGAN_HALF and DEVICE.type == "cuda", gpu_id=0 if DEVICE.type == "cuda" else None)
    def _realsr(img):
        bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        out_bgr, _ = upsampler.enhance(bgr, outscale=SR_REALESRGAN_OUTSCALE)
        return cv2.cvtColor(out_bgr, cv2.COLOR_BGR2RGB)
    return _realsr

# ============================
# MODEL FACTORY
# ============================

def build_model_and_preprocess():
    if MODEL_BACKBONE == "deeplab":
        net = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES, output_stride=OUTPUT_STRIDE, pretrained_backbone=False)
        net.load_state_dict(torch.load(DEEPLAB_CHECKPOINT, map_location=DEVICE, weights_only=False)["model_state"], strict=False)
        preprocess = T.Compose([T.ToTensor(), T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])])
        def _prep(pil):
            return preprocess(pil.convert("RGB"))
        def _fwd(t):
            with torch.no_grad():
                return net(t)
        return net.to(DEVICE).eval(), _prep, _fwd

    extractor = SegformerFeatureExtractor.from_pretrained(SEGFORMER_MODEL_NAME)
    net = SegformerForSemanticSegmentation.from_pretrained(SEGFORMER_MODEL_NAME).to(DEVICE).eval()
    def _fwd(pil_batch):
        inp = extractor(images=pil_batch, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            out = net(**inp).logits
        bs, _, h4, w4 = out.shape; h, w = h4*4, w4*4
        return torch.nn.functional.interpolate(out, size=(h,w), mode="bilinear", align_corners=False)
    return net, None, _fwd, extractor

# ============================
# OUTPUT FILE NAME
# ============================
_run = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
OUTPUT_JSON = f"metrics_{MODEL_BACKBONE}_{SR_BACKEND or 'nosr'}_thr{HIGH_CONF_THRESH}_rev{LOW_CONF_THRESH}_bs{BATCH_SIZE}_{_run}.json"

RUN_PARAMS = {
    "high_conf_thresh": HIGH_CONF_THRESH,
    "low_conf_thresh": LOW_CONF_THRESH,
    "morph_kernel": MORPH_KERNEL_SIZE,
    "model_backbone": MODEL_BACKBONE,
    "sr_backend": SR_BACKEND,
    "run_time": _run,
}

# ============================
# MAIN
# ============================

def main():
    if MODEL_BACKBONE == "segformer":
        net, _pre, forward, extractor = build_model_and_preprocess()
    else:
        net, preprocess, forward = build_model_and_preprocess()  # type: ignore

    model_splice = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES,
                                        output_stride=OUTPUT_STRIDE,
                                        pretrained_backbone=False)
    ckpt_splice = torch.load(CHECKPOINT_PATH_SPLICE, map_location=DEVICE, weights_only=False)
    model_splice.load_state_dict(ckpt_splice["model_state"], strict=False)
    model_splice.eval().to(DEVICE)

    sr = build_super_resolver()
    ds = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
    loader = DataLoader(ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=lambda b: tuple(zip(*b)))

    log: List[Dict[str,Any]] = []
    idx = 0

    for imgs, masks in tqdm(loader, desc="Inference", unit="batch"):
        if MODEL_BACKBONE == "segformer":
            logits_batch = forward(list(imgs))
        else:
            batch_t = torch.stack([preprocess(p) for p in imgs]).to(DEVICE)  # type: ignore[arg-type]
            logits_batch = forward(batch_t)

        for i in range(len(imgs)):
            img_np = np.array(imgs[i]); gt_np = np.vectorize(ID2TRAINID.get)(np.array(masks[i]))
            logits = logits_batch[i]
            if logits.shape[1:] != img_np.shape[:2]:
                logits = torch.nn.functional.interpolate(logits.unsqueeze(0), size=img_np.shape[:2], mode="bilinear", align_corners=False)[0]

            logits_np = (logits - logits.min()).div(logits.max()-logits.min()).cpu().numpy()
            max_prob = logits_np.max(0)

            # forward/high pred map
            high_mask = max_prob >= HIGH_CONF_THRESH
            logits_high = logits_np * high_mask
            high_pred = logits_high.argmax(0).astype(np.uint8)

            # reverse/low pred map
            low_mask = max_prob <= LOW_CONF_THRESH
            logits_low = logits_np * low_mask
            low_pred = logits_low.argmax(0).astype(np.uint8)

            closed = closing(high_pred)
            opened = opening(closed)
            refined = opened.copy()

            # small‑object SR refine
            for cid in SMALL_OBJECT_CLASSES:
                cls_m = (opened == cid).astype(np.uint8)
                contours,_ = cv2.findContours(cls_m, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                for cnt in contours:
                    x,y,w,h = cv2.boundingRect(cnt)
                    if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE: continue
                    crop = crop_with_padding(img_np,x,y,w,h)
                    sr_crop = sr(crop)
                    if MODEL_BACKBONE == "segformer":
                        inp = extractor(images=Image.fromarray(sr_crop), return_tensors="pt").to(DEVICE)
                        with torch.no_grad(): crop_logits = net(**inp).logits
                        crop_logits = torch.nn.functional.interpolate(crop_logits, size=sr_crop.shape[:2], mode="bilinear", align_corners=False)[0]
                    else:
                        ct = T.ToTensor()(sr_crop).unsqueeze(0)
                        ct = T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])(ct).to(DEVICE)
                        with torch.no_grad(): crop_logits = model_splice(ct)[0]
                    labels = torch.softmax(crop_logits,0).cpu().numpy().argmax(0)
                    labels = cv2.resize(labels,(w,h),interpolation=cv2.INTER_NEAREST)
                    refined[y:y+h, x:x+w] = labels

            raw_pred = logits_np.argmax(0).astype(np.uint8)

            log.append({
                "idx": idx,
                "iou_raw": calculate_iou_all(raw_pred, gt_np),
                "iou_high": calculate_iou_all(high_pred, gt_np),
                "iou_low": calculate_iou_all(low_pred, gt_np),
                "iou_closed": calculate_iou_all(closed, gt_np),
                "iou_opened": calculate_iou_all(opened, gt_np),
                "iou_refined": calculate_iou_all(refined, gt_np)
            })
            idx += 1

    RUN_PARAMS["finish_time"] = datetime.datetime.now().strftime("%Y-%m-%d_%H:%M:%S")
    with open(OUTPUT_JSON,"w",encoding="utf-8") as f:
        json.dump({"run_parameters": RUN_PARAMS, "metrics": log}, f, indent=2)

    print(f"Saved run → {OUTPUT_JSON}")

if __name__ == "__main__":
    main()
```

```python
"""Cityscapes semantic‑segmentation evaluation – batched, multi‑backbone (DeepLabV3+
| SegFormer), optional super‑resolution, *reverse+forward* confidence screening,
*morphological closing → opening* cleanup, per‑class metrics, self‑describing
JSON output.

What’s new (May 2025 rev)
-------------------------
* **Two‑sided confidence thresholding**
  * `HIGH_CONF_THRESH` keeps strong predictions (as before).
  * `LOW_CONF_THRESH` isolates *low*‑confidence pixels ("reverse threshold").
  * Both maps are logged for analysis.
* **Morphological opening after closing** to remove small islands once holes
  are filled.
* Metrics now include IoU for the *opened* mask and the isolated low‑conf mask.
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"

# ---- Backbone -----------------------------------------------------------
MODEL_BACKBONE: str = "deeplab"          # 'deeplab' | 'segformer'
DEEPLAB_CHECKPOINT: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"
SEGFORMER_MODEL_NAME: str = "nvidia/segformer-b1-finetuned-cityscapes-1024-1024"
OUTPUT_STRIDE: int = 16                  # only for DeepLab

NUM_CLASSES: int = 19
BATCH_SIZE: int = 4
SMALL_OBJECT_CLASSES: list[int] = [6, 7]
HIGH_CONF_THRESH: float = 0.5            # forward threshold
LOW_CONF_THRESH: float = 0.4             # reverse threshold
MORPH_KERNEL_SIZE: int = 20              # for closing & opening
MIN_INSTANCE_SIDE: int = 50

# ---- Super‑resolution ----------------------------------------------------
SR_BACKEND: str | None = "swin2sr"        # 'swin2sr' | 'realesrgan' | None
SR_MODEL_NAME: str = "caidas/swin2SR-classical-sr-x2-64"
SR_REALESRGAN_MODEL: str = "RealESRGAN_x4plus"
SR_REALESRGAN_DENOISE: float = 0.5
SR_REALESRGAN_TILE: int = 0
SR_REALESRGAN_TILE_PAD: int = 10
SR_REALESRGAN_OUTSCALE: int = 4
SR_REALESRGAN_HALF: bool = True

# ---- Runtime ------------------------------------------------------------
NUM_WORKERS: int = 4
SEED: int = 42

# ============================
# ENV & DEPENDENCIES
# ============================
import os, random, json, datetime
from pathlib import Path
from typing import List, Dict, Tuple, Any, Optional

import cv2, numpy as np, torch
from torch.utils.data import DataLoader
from torchvision import transforms as T
from tqdm.auto import tqdm
from PIL import Image

# visuals (comment‑out callers by default)
import matplotlib.pyplot as plt  # noqa: F401
import matplotlib.patches as mpatches  # noqa: F401

# Backbone‑specific ---------------------------------------
if MODEL_BACKBONE == "deeplab":
    from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
elif MODEL_BACKBONE == "segformer":
    from transformers import SegformerFeatureExtractor, SegformerForSemanticSegmentation
else:
    raise ValueError("Unsupported MODEL_BACKBONE")

# Super‑res imports --------------------------------------
if SR_BACKEND == "swin2sr":
    from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
elif SR_BACKEND == "realesrgan":
    from realesrgan import RealESRGANer
    from basicsr.archs.rrdbnet_arch import RRDBNet
    from basicsr.archs.srvgg_arch import SRVGGNetCompact
    from basicsr.utils.download_util import load_file_from_url

# Dataset meta -------------------------------------------
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes

CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# Device & RNG -------------------------------------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
if DEVICE.type == "cuda":
    torch.cuda.manual_seed_all(SEED)

# ============================
# UTILITIES
# ============================

def calculate_iou_all(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> List[Optional[float]]:
    ious: List[Optional[float]] = []
    for c in range(num_classes):
        inter = np.logical_and(pred == c, true == c).sum()
        union = np.logical_or(pred == c, true == c).sum()
        ious.append(inter / union if union else None)
    return ious


def morph(mask: np.ndarray, op: int, ksize: int = MORPH_KERNEL_SIZE):
    kernel = np.ones((ksize, ksize), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), op, kernel)

closing = lambda m: morph(m, cv2.MORPH_CLOSE)
opening = lambda m: morph(m, cv2.MORPH_OPEN)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    px, py = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - px), max(0, y - py)
    x2, y2 = min(img.shape[1], x + w + px), min(img.shape[0], y + h + py)
    return img[y1:y2, x1:x2]

# ============================
# SUPER‑RESOLUTION FACTORY
# ============================

def build_super_resolver():
    if SR_BACKEND is None:
        return lambda img: img

    if SR_BACKEND == "swin2sr":
        proc = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
        sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()
        def _swin(img):
            t = proc(img, return_tensors="pt").to(DEVICE)
            with torch.no_grad():
                out = sr_model(**t).reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
            return (np.moveaxis(out, 0, -1) * 255).round().astype(np.uint8)
        return _swin

    # Real‑ESRGAN
    def _realesr_arch(name):
        if name == "RealESRGAN_x4plus":
            return RRDBNet(3, 3, 4, 64, 23, 32), 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth"]
        if name == "RealESRGAN_x2plus":
            return RRDBNet(3, 3, 2, 64, 23, 32), 2, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.1/RealESRGAN_x2plus.pth"]
        return SRVGGNetCompact(3, 3, 64, num_conv=32, upscale=4, act_type="prelu"), 4, [
            "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-general-x4v3.pth"]

    arch, netscale, urls = _realesr_arch(SR_REALESRGAN_MODEL)
    wdir = Path("weights"); wdir.mkdir(exist_ok=True)
    model_paths = [str(load_file_from_url(u, model_dir=str(wdir), progress=True)) for u in urls]
    upsampler = RealESRGANer(scale=netscale, model_path=model_paths[0] if len(model_paths) == 1 else model_paths,
                             model=arch, tile=SR_REALESRGAN_TILE, tile_pad=SR_REALESRGAN_TILE_PAD,
                             half=SR_REALESRGAN_HALF and DEVICE.type == "cuda", gpu_id=0 if DEVICE.type == "cuda" else None)
    def _realsr(img):
        bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        out_bgr, _ = upsampler.enhance(bgr, outscale=SR_REALESRGAN_OUTSCALE)
        return cv2.cvtColor(out_bgr, cv2.COLOR_BGR2RGB)
    return _realsr

# ============================
# MODEL FACTORY
# ============================

def build_model_and_preprocess():
    if MODEL_BACKBONE == "deeplab":
        net = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES, output_stride=OUTPUT_STRIDE, pretrained_backbone=False)
        net.load_state_dict(torch.load(DEEPLAB_CHECKPOINT, map_location=DEVICE, weights_only=False)["model_state"], strict=False)
        preprocess = T.Compose([T.ToTensor(), T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])])
        def _prep(pil):
            return preprocess(pil.convert("RGB"))
        def _fwd(t):
            with torch.no_grad():
                return net(t)
        return net.to(DEVICE).eval(), _prep, _fwd

    extractor = SegformerFeatureExtractor.from_pretrained(SEGFORMER_MODEL_NAME)
    net = SegformerForSemanticSegmentation.from_pretrained(SEGFORMER_MODEL_NAME).to(DEVICE).eval()
    def _fwd(pil_batch):
        inp = extractor(images=pil_batch, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            out = net(**inp).logits
        bs, _, h4, w4 = out.shape; h, w = h4*4, w4*4
        return torch.nn.functional.interpolate(out, size=(h,w), mode="bilinear", align_corners=False)
    return net, None, _fwd, extractor

# ============================
# OUTPUT FILE NAME
# ============================
_run = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
OUTPUT_JSON = f"metrics_{MODEL_BACKBONE}_{SR_BACKEND or 'nosr'}_thr{HIGH_CONF_THRESH}_rev{LOW_CONF_THRESH}_bs{BATCH_SIZE}_{_run}.json"

RUN_PARAMS = {
    "high_conf_thresh": HIGH_CONF_THRESH,
    "low_conf_thresh": LOW_CONF_THRESH,
    "morph_kernel": MORPH_KERNEL_SIZE,
    "model_backbone": MODEL_BACKBONE,
    "sr_backend": SR_BACKEND,
    "run_time": _run,
}

# ============================
# MAIN
# ============================

def main():
    if MODEL_BACKBONE == "segformer":
        net, _pre, forward, extractor = build_model_and_preprocess()
    else:
        net, preprocess, forward = build_model_and_preprocess()  # type: ignore

    sr = build_super_resolver()
    ds = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
    loader = DataLoader(ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=lambda b: tuple(zip(*b)))

    log: List[Dict[str,Any]] = []
    idx = 0

    for imgs, masks in tqdm(loader, desc="Inference", unit="batch"):
        if MODEL_BACKBONE == "segformer":
            logits_batch = forward(list(imgs))
        else:
            batch_t = torch.stack([preprocess(p) for p in imgs]).to(DEVICE)  # type: ignore[arg-type]
            logits_batch = forward(batch_t)

        for i in range(len(imgs)):
            img_np = np.array(imgs[i]); gt_np = np.vectorize(ID2TRAINID.get)(np.array(masks[i]))
            logits = logits_batch[i]
            if logits.shape[1:] != img_np.shape[:2]:
                logits = torch.nn.functional.interpolate(logits.unsqueeze(0), size=img_np.shape[:2], mode="bilinear", align_corners=False)[0]

            logits_np = (logits - logits.min()).div(logits.max()-logits.min()).cpu().numpy()
            max_prob = logits_np.max(0)

            # forward/high pred map
            high_mask = max_prob >= HIGH_CONF_THRESH
            logits_high = logits_np * high_mask
            high_pred = logits_high.argmax(0).astype(np.uint8)

            # reverse/low pred map
            low_mask = max_prob <= LOW_CONF_THRESH
            logits_low = logits_np * low_mask
            low_pred = logits_low.argmax(0).astype(np.uint8)

            closed = closing(high_pred)
            opened = opening(closed)
            refined = opened.copy()

            # small‑object SR refine
            for cid in SMALL_OBJECT_CLASSES:
                cls_m = (opened == cid).astype(np.uint8)
                contours,_ = cv2.findContours(cls_m, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                for cnt in contours:
                    x,y,w,h = cv2.boundingRect(cnt)
                    if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE: continue
                    crop = crop_with_padding(img_np,x,y,w,h)
                    sr_crop = sr(crop)
                    if MODEL_BACKBONE == "segformer":
                        inp = extractor(images=Image.fromarray(sr_crop), return_tensors="pt").to(DEVICE)
                        with torch.no_grad(): crop_logits = net(**inp).logits
                        crop_logits = torch.nn.functional.interpolate(crop_logits, size=sr_crop.shape[:2], mode="bilinear", align_corners=False)[0]
                    else:
                        ct = T.ToTensor()(sr_crop).unsqueeze(0)
                        ct = T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])(ct).to(DEVICE)
                        with torch.no_grad(): crop_logits = net(ct)[0]
                    labels = torch.softmax(crop_logits,0).cpu().numpy().argmax(0)
                    labels = cv2.resize(labels,(w,h),interpolation=cv2.INTER_NEAREST)
                    refined[y:y+h, x:x+w] = labels

            raw_pred = logits_np.argmax(0).astype(np.uint8)

            log.append({
                "idx": idx,
                "iou_raw": calculate_iou_all(raw_pred, gt_np),
                "iou_high": calculate_iou_all(high_pred, gt_np),
                "iou_low": calculate_iou_all(low_pred, gt_np),
                "iou_closed": calculate_iou_all(closed, gt_np),
                "iou_opened": calculate_iou_all(opened, gt_np),
                "iou_refined": calculate_iou_all(refined, gt_np)
            })
            idx += 1

    RUN_PARAMS["finish_time"] = datetime.datetime.now().strftime("%Y-%m-%d_%H:%M:%S")
    with open(OUTPUT_JSON,"w",encoding="utf-8") as f:
        json.dump({"run_parameters": RUN_PARAMS, "metrics": log}, f, indent=2)

    print(f"Saved run → {OUTPUT_JSON}")

if __name__ == "__main__":
    main()
```

```python
"""Cityscapes semantic‑segmentation evaluation – batched, multi‑backbone (DeepLabV3+
| SegFormer), optional super‑resolution, *reverse+forward* confidence screening,
*morphological closing → opening* cleanup, per‑class metrics, self‑describing
JSON output.

What’s new (May 2025 rev)
-------------------------
* **Two‑sided confidence thresholding**
  * `HIGH_CONF_THRESH` keeps strong predictions (as before).
  * `LOW_CONF_THRESH` isolates *low*‑confidence pixels ("reverse threshold").
  * Both maps are logged for analysis.
* **Morphological opening after closing** to remove small islands once holes
  are filled.
* Metrics now include IoU for the *opened* mask and the isolated low‑conf mask.
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"

# ---- Backbone -----------------------------------------------------------
MODEL_BACKBONE: str = "segformer"          # 'deeplab' | 'segformer'
DEEPLAB_CHECKPOINT: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"
SEGFORMER_MODEL_NAME: str = "nvidia/segformer-b1-finetuned-cityscapes-1024-1024"
OUTPUT_STRIDE: int = 16                  # only for DeepLab

NUM_CLASSES: int = 19
BATCH_SIZE: int = 4
SMALL_OBJECT_CLASSES: list[int] = [6, 7]
HIGH_CONF_THRESH: float = 0.5            # forward threshold
LOW_CONF_THRESH: float = 0.4             # reverse threshold
MORPH_KERNEL_SIZE: int = 20              # for closing & opening
MIN_INSTANCE_SIDE: int = 50

# ---- Super‑resolution ----------------------------------------------------
SR_BACKEND: str | None = "realesrgan"        # 'swin2sr' | 'realesrgan' | None
SR_MODEL_NAME: str = "caidas/swin2SR-classical-sr-x2-64"
SR_REALESRGAN_MODEL: str = "RealESRGAN_x4plus"
SR_REALESRGAN_DENOISE: float = 0.5
SR_REALESRGAN_TILE: int = 0
SR_REALESRGAN_TILE_PAD: int = 10
SR_REALESRGAN_OUTSCALE: int = 4
SR_REALESRGAN_HALF: bool = True

# ---- Runtime ------------------------------------------------------------
NUM_WORKERS: int = 4
SEED: int = 42

# ============================
# ENV & DEPENDENCIES
# ============================
import os, random, json, datetime
from pathlib import Path
from typing import List, Dict, Tuple, Any, Optional

import cv2, numpy as np, torch
from torch.utils.data import DataLoader
from torchvision import transforms as T
from tqdm.auto import tqdm
from PIL import Image

# visuals (comment‑out callers by default)
import matplotlib.pyplot as plt  # noqa: F401
import matplotlib.patches as mpatches  # noqa: F401

# Backbone‑specific ---------------------------------------
if MODEL_BACKBONE == "deeplab":
    from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
elif MODEL_BACKBONE == "segformer":
    from transformers import SegformerFeatureExtractor, SegformerForSemanticSegmentation
else:
    raise ValueError("Unsupported MODEL_BACKBONE")

# Super‑res imports --------------------------------------
if SR_BACKEND == "swin2sr":
    from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
elif SR_BACKEND == "realesrgan":
    from realesrgan import RealESRGANer
    from basicsr.archs.rrdbnet_arch import RRDBNet
    from basicsr.archs.srvgg_arch import SRVGGNetCompact
    from basicsr.utils.download_util import load_file_from_url

# Dataset meta -------------------------------------------
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes

CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# Device & RNG -------------------------------------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
if DEVICE.type == "cuda":
    torch.cuda.manual_seed_all(SEED)

# ============================
# UTILITIES
# ============================

def calculate_iou_all(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> List[Optional[float]]:
    ious: List[Optional[float]] = []
    for c in range(num_classes):
        inter = np.logical_and(pred == c, true == c).sum()
        union = np.logical_or(pred == c, true == c).sum()
        ious.append(inter / union if union else None)
    return ious


def morph(mask: np.ndarray, op: int, ksize: int = MORPH_KERNEL_SIZE):
    kernel = np.ones((ksize, ksize), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), op, kernel)

closing = lambda m: morph(m, cv2.MORPH_CLOSE)
opening = lambda m: morph(m, cv2.MORPH_OPEN)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    px, py = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - px), max(0, y - py)
    x2, y2 = min(img.shape[1], x + w + px), min(img.shape[0], y + h + py)
    return img[y1:y2, x1:x2]

# ============================
# SUPER‑RESOLUTION FACTORY
# ============================

def build_super_resolver():
    if SR_BACKEND is None:
        return lambda img: img

    if SR_BACKEND == "swin2sr":
        proc = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
        sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()
        def _swin(img):
            t = proc(img, return_tensors="pt").to(DEVICE)
            with torch.no_grad():
                out = sr_model(**t).reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
            return (np.moveaxis(out, 0, -1) * 255).round().astype(np.uint8)
        return _swin

    # Real‑ESRGAN
    def _realesr_arch(name):
        if name == "RealESRGAN_x4plus":
            return RRDBNet(3, 3, 4, 64, 23, 32), 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth"]
        if name == "RealESRGAN_x2plus":
            return RRDBNet(3, 3, 2, 64, 23, 32), 2, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.1/RealESRGAN_x2plus.pth"]
        return SRVGGNetCompact(3, 3, 64, num_conv=32, upscale=4, act_type="prelu"), 4, [
            "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-general-x4v3.pth"]

    arch, netscale, urls = _realesr_arch(SR_REALESRGAN_MODEL)
    wdir = Path("weights"); wdir.mkdir(exist_ok=True)
    model_paths = [str(load_file_from_url(u, model_dir=str(wdir), progress=True)) for u in urls]
    upsampler = RealESRGANer(scale=netscale, model_path=model_paths[0] if len(model_paths) == 1 else model_paths,
                             model=arch, tile=SR_REALESRGAN_TILE, tile_pad=SR_REALESRGAN_TILE_PAD,
                             half=SR_REALESRGAN_HALF and DEVICE.type == "cuda", gpu_id=0 if DEVICE.type == "cuda" else None)
    def _realsr(img):
        bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        out_bgr, _ = upsampler.enhance(bgr, outscale=SR_REALESRGAN_OUTSCALE)
        return cv2.cvtColor(out_bgr, cv2.COLOR_BGR2RGB)
    return _realsr

# ============================
# MODEL FACTORY
# ============================

def build_model_and_preprocess():
    if MODEL_BACKBONE == "deeplab":
        net = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES, output_stride=OUTPUT_STRIDE, pretrained_backbone=False)
        net.load_state_dict(torch.load(DEEPLAB_CHECKPOINT, map_location=DEVICE, weights_only=False)["model_state"], strict=False)
        preprocess = T.Compose([T.ToTensor(), T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])])
        def _prep(pil):
            return preprocess(pil.convert("RGB"))
        def _fwd(t):
            with torch.no_grad():
                return net(t)
        return net.to(DEVICE).eval(), _prep, _fwd

    extractor = SegformerFeatureExtractor.from_pretrained(SEGFORMER_MODEL_NAME)
    net = SegformerForSemanticSegmentation.from_pretrained(SEGFORMER_MODEL_NAME).to(DEVICE).eval()
    def _fwd(pil_batch):
        inp = extractor(images=pil_batch, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            out = net(**inp).logits
        bs, _, h4, w4 = out.shape; h, w = h4*4, w4*4
        return torch.nn.functional.interpolate(out, size=(h,w), mode="bilinear", align_corners=False)
    return net, None, _fwd, extractor

# ============================
# OUTPUT FILE NAME
# ============================
_run = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
OUTPUT_JSON = f"metrics_{MODEL_BACKBONE}_{SR_BACKEND or 'nosr'}_thr{HIGH_CONF_THRESH}_rev{LOW_CONF_THRESH}_bs{BATCH_SIZE}_{_run}.json"

RUN_PARAMS = {
    "high_conf_thresh": HIGH_CONF_THRESH,
    "low_conf_thresh": LOW_CONF_THRESH,
    "morph_kernel": MORPH_KERNEL_SIZE,
    "model_backbone": MODEL_BACKBONE,
    "sr_backend": SR_BACKEND,
    "run_time": _run,
}

# ============================
# MAIN
# ============================

def main():
    if MODEL_BACKBONE == "segformer":
        net, _pre, forward, extractor = build_model_and_preprocess()
    else:
        net, preprocess, forward = build_model_and_preprocess()  # type: ignore

    sr = build_super_resolver()
    ds = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
    loader = DataLoader(ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=lambda b: tuple(zip(*b)))

    log: List[Dict[str,Any]] = []
    idx = 0

    for imgs, masks in tqdm(loader, desc="Inference", unit="batch"):
        if MODEL_BACKBONE == "segformer":
            logits_batch = forward(list(imgs))
        else:
            batch_t = torch.stack([preprocess(p) for p in imgs]).to(DEVICE)  # type: ignore[arg-type]
            logits_batch = forward(batch_t)

        for i in range(len(imgs)):
            img_np = np.array(imgs[i]); gt_np = np.vectorize(ID2TRAINID.get)(np.array(masks[i]))
            logits = logits_batch[i]
            if logits.shape[1:] != img_np.shape[:2]:
                logits = torch.nn.functional.interpolate(logits.unsqueeze(0), size=img_np.shape[:2], mode="bilinear", align_corners=False)[0]

            logits_np = (logits - logits.min()).div(logits.max()-logits.min()).cpu().numpy()
            max_prob = logits_np.max(0)

            # forward/high pred map
            high_mask = max_prob >= HIGH_CONF_THRESH
            logits_high = logits_np * high_mask
            high_pred = logits_high.argmax(0).astype(np.uint8)

            # reverse/low pred map
            low_mask = max_prob <= LOW_CONF_THRESH
            logits_low = logits_np * low_mask
            low_pred = logits_low.argmax(0).astype(np.uint8)

            closed = closing(high_pred)
            opened = opening(closed)
            refined = opened.copy()

            # small‑object SR refine
            for cid in SMALL_OBJECT_CLASSES:
                cls_m = (opened == cid).astype(np.uint8)
                contours,_ = cv2.findContours(cls_m, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                for cnt in contours:
                    x,y,w,h = cv2.boundingRect(cnt)
                    if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE: continue
                    crop = crop_with_padding(img_np,x,y,w,h)
                    sr_crop = sr(crop)
                    if MODEL_BACKBONE == "segformer":
                        inp = extractor(images=Image.fromarray(sr_crop), return_tensors="pt").to(DEVICE)
                        with torch.no_grad(): crop_logits = net(**inp).logits
                        crop_logits = torch.nn.functional.interpolate(crop_logits, size=sr_crop.shape[:2], mode="bilinear", align_corners=False)[0]
                    else:
                        ct = T.ToTensor()(sr_crop).unsqueeze(0)
                        ct = T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])(ct).to(DEVICE)
                        with torch.no_grad(): crop_logits = net(ct)[0]
                    labels = torch.softmax(crop_logits,0).cpu().numpy().argmax(0)
                    labels = cv2.resize(labels,(w,h),interpolation=cv2.INTER_NEAREST)
                    refined[y:y+h, x:x+w] = labels

            raw_pred = logits_np.argmax(0).astype(np.uint8)

            log.append({
                "idx": idx,
                "iou_raw": calculate_iou_all(raw_pred, gt_np),
                "iou_high": calculate_iou_all(high_pred, gt_np),
                "iou_low": calculate_iou_all(low_pred, gt_np),
                "iou_closed": calculate_iou_all(closed, gt_np),
                "iou_opened": calculate_iou_all(opened, gt_np),
                "iou_refined": calculate_iou_all(refined, gt_np)
            })
            idx += 1

    RUN_PARAMS["finish_time"] = datetime.datetime.now().strftime("%Y-%m-%d_%H:%M:%S")
    with open(OUTPUT_JSON,"w",encoding="utf-8") as f:
        json.dump({"run_parameters": RUN_PARAMS, "metrics": log}, f, indent=2)

    print(f"Saved run → {OUTPUT_JSON}")

if __name__ == "__main__":
    main()
```

```python
"""Cityscapes semantic‑segmentation evaluation – batched, multi‑backbone (DeepLabV3+
| SegFormer), optional super‑resolution, *reverse+forward* confidence screening,
*morphological closing → opening* cleanup, per‑class metrics, self‑describing
JSON output.

What’s new (May 2025 rev)
-------------------------
* **Two‑sided confidence thresholding**
  * `HIGH_CONF_THRESH` keeps strong predictions (as before).
  * `LOW_CONF_THRESH` isolates *low*‑confidence pixels ("reverse threshold").
  * Both maps are logged for analysis.
* **Morphological opening after closing** to remove small islands once holes
  are filled.
* Metrics now include IoU for the *opened* mask and the isolated low‑conf mask.
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"

# ---- Backbone -----------------------------------------------------------
MODEL_BACKBONE: str = "segformer"          # 'deeplab' | 'segformer'
DEEPLAB_CHECKPOINT: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"
SEGFORMER_MODEL_NAME: str = "nvidia/segformer-b1-finetuned-cityscapes-1024-1024"
OUTPUT_STRIDE: int = 16                  # only for DeepLab

NUM_CLASSES: int = 19
BATCH_SIZE: int = 4
SMALL_OBJECT_CLASSES: list[int] = [6, 7]
HIGH_CONF_THRESH: float = 0.5            # forward threshold
LOW_CONF_THRESH: float = 0.4             # reverse threshold
MORPH_KERNEL_SIZE: int = 20              # for closing & opening
MIN_INSTANCE_SIDE: int = 50

# ---- Super‑resolution ----------------------------------------------------
SR_BACKEND: str | None = "realesrgan"        # 'swin2sr' | 'realesrgan' | None
SR_MODEL_NAME: str = "caidas/swin2SR-classical-sr-x2-64"
SR_REALESRGAN_MODEL: str = "RealESRGAN_x2plus"
SR_REALESRGAN_DENOISE: float = 0.5
SR_REALESRGAN_TILE: int = 0
SR_REALESRGAN_TILE_PAD: int = 10
SR_REALESRGAN_OUTSCALE: int = 2
SR_REALESRGAN_HALF: bool = True

# ---- Runtime ------------------------------------------------------------
NUM_WORKERS: int = 4
SEED: int = 42

# ============================
# ENV & DEPENDENCIES
# ============================
import os, random, json, datetime
from pathlib import Path
from typing import List, Dict, Tuple, Any, Optional

import cv2, numpy as np, torch
from torch.utils.data import DataLoader
from torchvision import transforms as T
from tqdm.auto import tqdm
from PIL import Image

# visuals (comment‑out callers by default)
import matplotlib.pyplot as plt  # noqa: F401
import matplotlib.patches as mpatches  # noqa: F401

# Backbone‑specific ---------------------------------------
if MODEL_BACKBONE == "deeplab":
    from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
elif MODEL_BACKBONE == "segformer":
    from transformers import SegformerFeatureExtractor, SegformerForSemanticSegmentation
else:
    raise ValueError("Unsupported MODEL_BACKBONE")

# Super‑res imports --------------------------------------
if SR_BACKEND == "swin2sr":
    from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
elif SR_BACKEND == "realesrgan":
    from realesrgan import RealESRGANer
    from basicsr.archs.rrdbnet_arch import RRDBNet
    from basicsr.archs.srvgg_arch import SRVGGNetCompact
    from basicsr.utils.download_util import load_file_from_url

# Dataset meta -------------------------------------------
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes

CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# Device & RNG -------------------------------------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
if DEVICE.type == "cuda":
    torch.cuda.manual_seed_all(SEED)

# ============================
# UTILITIES
# ============================

def calculate_iou_all(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> List[Optional[float]]:
    ious: List[Optional[float]] = []
    for c in range(num_classes):
        inter = np.logical_and(pred == c, true == c).sum()
        union = np.logical_or(pred == c, true == c).sum()
        ious.append(inter / union if union else None)
    return ious


def morph(mask: np.ndarray, op: int, ksize: int = MORPH_KERNEL_SIZE):
    kernel = np.ones((ksize, ksize), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), op, kernel)

closing = lambda m: morph(m, cv2.MORPH_CLOSE)
opening = lambda m: morph(m, cv2.MORPH_OPEN)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    px, py = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - px), max(0, y - py)
    x2, y2 = min(img.shape[1], x + w + px), min(img.shape[0], y + h + py)
    return img[y1:y2, x1:x2]

# ============================
# SUPER‑RESOLUTION FACTORY
# ============================

def build_super_resolver():
    if SR_BACKEND is None:
        return lambda img: img

    if SR_BACKEND == "swin2sr":
        proc = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
        sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()
        def _swin(img):
            t = proc(img, return_tensors="pt").to(DEVICE)
            with torch.no_grad():
                out = sr_model(**t).reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
            return (np.moveaxis(out, 0, -1) * 255).round().astype(np.uint8)
        return _swin

    # Real‑ESRGAN
    def _realesr_arch(name):
        if name == "RealESRGAN_x4plus":
            return RRDBNet(3, 3, 4, 64, 23, 32), 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth"]
        if name == "RealESRGAN_x2plus":
            return RRDBNet(3, 3, 2, 64, 23, 32), 2, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.1/RealESRGAN_x2plus.pth"]
        return SRVGGNetCompact(3, 3, 64, num_conv=32, upscale=4, act_type="prelu"), 4, [
            "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-general-x4v3.pth"]

    arch, netscale, urls = _realesr_arch(SR_REALESRGAN_MODEL)
    wdir = Path("weights"); wdir.mkdir(exist_ok=True)
    model_paths = [str(load_file_from_url(u, model_dir=str(wdir), progress=True)) for u in urls]
    upsampler = RealESRGANer(scale=netscale, model_path=model_paths[0] if len(model_paths) == 1 else model_paths,
                             model=arch, tile=SR_REALESRGAN_TILE, tile_pad=SR_REALESRGAN_TILE_PAD,
                             half=SR_REALESRGAN_HALF and DEVICE.type == "cuda", gpu_id=0 if DEVICE.type == "cuda" else None)
    def _realsr(img):
        bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        out_bgr, _ = upsampler.enhance(bgr, outscale=SR_REALESRGAN_OUTSCALE)
        return cv2.cvtColor(out_bgr, cv2.COLOR_BGR2RGB)
    return _realsr

# ============================
# MODEL FACTORY
# ============================

def build_model_and_preprocess():
    if MODEL_BACKBONE == "deeplab":
        net = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES, output_stride=OUTPUT_STRIDE, pretrained_backbone=False)
        net.load_state_dict(torch.load(DEEPLAB_CHECKPOINT, map_location=DEVICE, weights_only=False)["model_state"], strict=False)
        preprocess = T.Compose([T.ToTensor(), T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])])
        def _prep(pil):
            return preprocess(pil.convert("RGB"))
        def _fwd(t):
            with torch.no_grad():
                return net(t)
        return net.to(DEVICE).eval(), _prep, _fwd

    extractor = SegformerFeatureExtractor.from_pretrained(SEGFORMER_MODEL_NAME)
    net = SegformerForSemanticSegmentation.from_pretrained(SEGFORMER_MODEL_NAME).to(DEVICE).eval()
    def _fwd(pil_batch):
        inp = extractor(images=pil_batch, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            out = net(**inp).logits
        bs, _, h4, w4 = out.shape; h, w = h4*4, w4*4
        return torch.nn.functional.interpolate(out, size=(h,w), mode="bilinear", align_corners=False)
    return net, None, _fwd, extractor

# ============================
# OUTPUT FILE NAME
# ============================
_run = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
OUTPUT_JSON = f"metrics_{MODEL_BACKBONE}_{SR_BACKEND or 'nosr'}_thr{HIGH_CONF_THRESH}_rev{LOW_CONF_THRESH}_bs{BATCH_SIZE}_{_run}.json"

RUN_PARAMS = {
    "high_conf_thresh": HIGH_CONF_THRESH,
    "low_conf_thresh": LOW_CONF_THRESH,
    "morph_kernel": MORPH_KERNEL_SIZE,
    "model_backbone": MODEL_BACKBONE,
    "sr_backend": SR_BACKEND,
    "run_time": _run,
}

# ============================
# MAIN
# ============================

def main():
    if MODEL_BACKBONE == "segformer":
        net, _pre, forward, extractor = build_model_and_preprocess()
    else:
        net, preprocess, forward = build_model_and_preprocess()  # type: ignore

    sr = build_super_resolver()
    ds = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
    loader = DataLoader(ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=lambda b: tuple(zip(*b)))

    log: List[Dict[str,Any]] = []
    idx = 0

    for imgs, masks in tqdm(loader, desc="Inference", unit="batch"):
        if MODEL_BACKBONE == "segformer":
            logits_batch = forward(list(imgs))
        else:
            batch_t = torch.stack([preprocess(p) for p in imgs]).to(DEVICE)  # type: ignore[arg-type]
            logits_batch = forward(batch_t)

        for i in range(len(imgs)):
            img_np = np.array(imgs[i]); gt_np = np.vectorize(ID2TRAINID.get)(np.array(masks[i]))
            logits = logits_batch[i]
            if logits.shape[1:] != img_np.shape[:2]:
                logits = torch.nn.functional.interpolate(logits.unsqueeze(0), size=img_np.shape[:2], mode="bilinear", align_corners=False)[0]

            logits_np = (logits - logits.min()).div(logits.max()-logits.min()).cpu().numpy()
            max_prob = logits_np.max(0)

            # forward/high pred map
            high_mask = max_prob >= HIGH_CONF_THRESH
            logits_high = logits_np * high_mask
            high_pred = logits_high.argmax(0).astype(np.uint8)

            # reverse/low pred map
            low_mask = max_prob <= LOW_CONF_THRESH
            logits_low = logits_np * low_mask
            low_pred = logits_low.argmax(0).astype(np.uint8)

            closed = closing(high_pred)
            opened = opening(closed)
            refined = opened.copy()

            # small‑object SR refine
            for cid in SMALL_OBJECT_CLASSES:
                cls_m = (opened == cid).astype(np.uint8)
                contours,_ = cv2.findContours(cls_m, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                for cnt in contours:
                    x,y,w,h = cv2.boundingRect(cnt)
                    if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE: continue
                    crop = crop_with_padding(img_np,x,y,w,h)
                    sr_crop = sr(crop)
                    if MODEL_BACKBONE == "segformer":
                        inp = extractor(images=Image.fromarray(sr_crop), return_tensors="pt").to(DEVICE)
                        with torch.no_grad(): crop_logits = net(**inp).logits
                        crop_logits = torch.nn.functional.interpolate(crop_logits, size=sr_crop.shape[:2], mode="bilinear", align_corners=False)[0]
                    else:
                        ct = T.ToTensor()(sr_crop).unsqueeze(0)
                        ct = T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])(ct).to(DEVICE)
                        with torch.no_grad(): crop_logits = net(ct)[0]
                    labels = torch.softmax(crop_logits,0).cpu().numpy().argmax(0)
                    labels = cv2.resize(labels,(w,h),interpolation=cv2.INTER_NEAREST)
                    refined[y:y+h, x:x+w] = labels

            raw_pred = logits_np.argmax(0).astype(np.uint8)

            log.append({
                "idx": idx,
                "iou_raw": calculate_iou_all(raw_pred, gt_np),
                "iou_high": calculate_iou_all(high_pred, gt_np),
                "iou_low": calculate_iou_all(low_pred, gt_np),
                "iou_closed": calculate_iou_all(closed, gt_np),
                "iou_opened": calculate_iou_all(opened, gt_np),
                "iou_refined": calculate_iou_all(refined, gt_np)
            })
            idx += 1

    RUN_PARAMS["finish_time"] = datetime.datetime.now().strftime("%Y-%m-%d_%H:%M:%S")
    with open(OUTPUT_JSON,"w",encoding="utf-8") as f:
        json.dump({"run_parameters": RUN_PARAMS, "metrics": log}, f, indent=2)

    print(f"Saved run → {OUTPUT_JSON}")

if __name__ == "__main__":
    main()
```

```python
"""Cityscapes semantic‑segmentation evaluation – batched, multi‑backbone (DeepLabV3+
| SegFormer), optional super‑resolution, *reverse+forward* confidence screening,
*morphological closing → opening* cleanup, per‑class metrics, self‑describing
JSON output.

What’s new (May 2025 rev)
-------------------------
* **Two‑sided confidence thresholding**
  * `HIGH_CONF_THRESH` keeps strong predictions (as before).
  * `LOW_CONF_THRESH` isolates *low*‑confidence pixels ("reverse threshold").
  * Both maps are logged for analysis.
* **Morphological opening after closing** to remove small islands once holes
  are filled.
* Metrics now include IoU for the *opened* mask and the isolated low‑conf mask.
"""

# ============================
# CONFIGURATION
# ============================
DATASET_ROOT: str = "./datasets/cityscapes"

# ---- Backbone -----------------------------------------------------------
MODEL_BACKBONE: str = "segformer"          # 'deeplab' | 'segformer'
DEEPLAB_CHECKPOINT: str = "./best_deeplabv3plus_mobilenet_cityscapes_os16.pth"
SEGFORMER_MODEL_NAME: str = "nvidia/segformer-b1-finetuned-cityscapes-1024-1024"
OUTPUT_STRIDE: int = 16                  # only for DeepLab

NUM_CLASSES: int = 19
BATCH_SIZE: int = 4
SMALL_OBJECT_CLASSES: list[int] = [6, 7]
HIGH_CONF_THRESH: float = 0.5            # forward threshold
LOW_CONF_THRESH: float = 0.4             # reverse threshold
MORPH_KERNEL_SIZE: int = 20              # for closing & opening
MIN_INSTANCE_SIDE: int = 50

# ---- Super‑resolution ----------------------------------------------------
SR_BACKEND: str | None = "swin2sr"        # 'swin2sr' | 'realesrgan' | None
SR_MODEL_NAME: str = "caidas/swin2SR-lightweight-x2-64"
SR_REALESRGAN_MODEL: str = "RealESRGAN_x2plus"
SR_REALESRGAN_DENOISE: float = 0.5
SR_REALESRGAN_TILE: int = 0
SR_REALESRGAN_TILE_PAD: int = 10
SR_REALESRGAN_OUTSCALE: int = 2
SR_REALESRGAN_HALF: bool = True

# ---- Runtime ------------------------------------------------------------
NUM_WORKERS: int = 4
SEED: int = 42

# ============================
# ENV & DEPENDENCIES
# ============================
import os, random, json, datetime
from pathlib import Path
from typing import List, Dict, Tuple, Any, Optional

import cv2, numpy as np, torch
from torch.utils.data import DataLoader
from torchvision import transforms as T
from tqdm.auto import tqdm
from PIL import Image

# visuals (comment‑out callers by default)
import matplotlib.pyplot as plt  # noqa: F401
import matplotlib.patches as mpatches  # noqa: F401

# Backbone‑specific ---------------------------------------
if MODEL_BACKBONE == "deeplab":
    from deeplabv3plus.network.modeling import deeplabv3plus_mobilenet
elif MODEL_BACKBONE == "segformer":
    from transformers import SegformerFeatureExtractor, SegformerForSemanticSegmentation
else:
    raise ValueError("Unsupported MODEL_BACKBONE")

# Super‑res imports --------------------------------------
if SR_BACKEND == "swin2sr":
    from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution
elif SR_BACKEND == "realesrgan":
    from realesrgan import RealESRGANer
    from basicsr.archs.rrdbnet_arch import RRDBNet
    from basicsr.archs.srvgg_arch import SRVGGNetCompact
    from basicsr.utils.download_util import load_file_from_url

# Dataset meta -------------------------------------------
from deeplabv3plus.datasets import Cityscapes as DeeplabCityscapes
from torchvision.datasets import Cityscapes

CITYSCAPES_LABELS = {c.train_id: c.name for c in DeeplabCityscapes.classes}
ID2TRAINID = {c.id: c.train_id for c in DeeplabCityscapes.classes}

# Device & RNG -------------------------------------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
if DEVICE.type == "cuda":
    torch.cuda.manual_seed_all(SEED)

# ============================
# UTILITIES
# ============================

def calculate_iou_all(pred: np.ndarray, true: np.ndarray, num_classes: int = NUM_CLASSES) -> List[Optional[float]]:
    ious: List[Optional[float]] = []
    for c in range(num_classes):
        inter = np.logical_and(pred == c, true == c).sum()
        union = np.logical_or(pred == c, true == c).sum()
        ious.append(inter / union if union else None)
    return ious


def morph(mask: np.ndarray, op: int, ksize: int = MORPH_KERNEL_SIZE):
    kernel = np.ones((ksize, ksize), np.uint8)
    return cv2.morphologyEx(mask.astype(np.uint8), op, kernel)

closing = lambda m: morph(m, cv2.MORPH_CLOSE)
opening = lambda m: morph(m, cv2.MORPH_OPEN)


def crop_with_padding(img: np.ndarray, x: int, y: int, w: int, h: int, pad: float = 0.1) -> np.ndarray:
    px, py = int(w * pad), int(h * pad)
    x1, y1 = max(0, x - px), max(0, y - py)
    x2, y2 = min(img.shape[1], x + w + px), min(img.shape[0], y + h + py)
    return img[y1:y2, x1:x2]

# ============================
# SUPER‑RESOLUTION FACTORY
# ============================

def build_super_resolver():
    if SR_BACKEND is None:
        return lambda img: img

    if SR_BACKEND == "swin2sr":
        proc = AutoImageProcessor.from_pretrained(SR_MODEL_NAME)
        sr_model = Swin2SRForImageSuperResolution.from_pretrained(SR_MODEL_NAME).to(DEVICE).eval()
        def _swin(img):
            t = proc(img, return_tensors="pt").to(DEVICE)
            with torch.no_grad():
                out = sr_model(**t).reconstruction.squeeze().float().clamp_(0, 1).cpu().numpy()
            return (np.moveaxis(out, 0, -1) * 255).round().astype(np.uint8)
        return _swin

    # Real‑ESRGAN
    def _realesr_arch(name):
        if name == "RealESRGAN_x4plus":
            return RRDBNet(3, 3, 4, 64, 23, 32), 4, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth"]
        if name == "RealESRGAN_x2plus":
            return RRDBNet(3, 3, 2, 64, 23, 32), 2, [
                "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.1/RealESRGAN_x2plus.pth"]
        return SRVGGNetCompact(3, 3, 64, num_conv=32, upscale=4, act_type="prelu"), 4, [
            "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-general-x4v3.pth"]

    arch, netscale, urls = _realesr_arch(SR_REALESRGAN_MODEL)
    wdir = Path("weights"); wdir.mkdir(exist_ok=True)
    model_paths = [str(load_file_from_url(u, model_dir=str(wdir), progress=True)) for u in urls]
    upsampler = RealESRGANer(scale=netscale, model_path=model_paths[0] if len(model_paths) == 1 else model_paths,
                             model=arch, tile=SR_REALESRGAN_TILE, tile_pad=SR_REALESRGAN_TILE_PAD,
                             half=SR_REALESRGAN_HALF and DEVICE.type == "cuda", gpu_id=0 if DEVICE.type == "cuda" else None)
    def _realsr(img):
        bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        out_bgr, _ = upsampler.enhance(bgr, outscale=SR_REALESRGAN_OUTSCALE)
        return cv2.cvtColor(out_bgr, cv2.COLOR_BGR2RGB)
    return _realsr

# ============================
# MODEL FACTORY
# ============================

def build_model_and_preprocess():
    if MODEL_BACKBONE == "deeplab":
        net = deeplabv3plus_mobilenet(num_classes=NUM_CLASSES, output_stride=OUTPUT_STRIDE, pretrained_backbone=False)
        net.load_state_dict(torch.load(DEEPLAB_CHECKPOINT, map_location=DEVICE, weights_only=False)["model_state"], strict=False)
        preprocess = T.Compose([T.ToTensor(), T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])])
        def _prep(pil):
            return preprocess(pil.convert("RGB"))
        def _fwd(t):
            with torch.no_grad():
                return net(t)
        return net.to(DEVICE).eval(), _prep, _fwd

    extractor = SegformerFeatureExtractor.from_pretrained(SEGFORMER_MODEL_NAME)
    net = SegformerForSemanticSegmentation.from_pretrained(SEGFORMER_MODEL_NAME).to(DEVICE).eval()
    def _fwd(pil_batch):
        inp = extractor(images=pil_batch, return_tensors="pt").to(DEVICE)
        with torch.no_grad():
            out = net(**inp).logits
        bs, _, h4, w4 = out.shape; h, w = h4*4, w4*4
        return torch.nn.functional.interpolate(out, size=(h,w), mode="bilinear", align_corners=False)
    return net, None, _fwd, extractor

# ============================
# OUTPUT FILE NAME
# ============================
_run = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
OUTPUT_JSON = f"metrics_{MODEL_BACKBONE}_{SR_BACKEND or 'nosr'}_thr{HIGH_CONF_THRESH}_rev{LOW_CONF_THRESH}_bs{BATCH_SIZE}_{_run}.json"

RUN_PARAMS = {
    "high_conf_thresh": HIGH_CONF_THRESH,
    "low_conf_thresh": LOW_CONF_THRESH,
    "morph_kernel": MORPH_KERNEL_SIZE,
    "model_backbone": MODEL_BACKBONE,
    "sr_backend": SR_BACKEND,
    "run_time": _run,
}

# ============================
# MAIN
# ============================

def main():
    if MODEL_BACKBONE == "segformer":
        net, _pre, forward, extractor = build_model_and_preprocess()
    else:
        net, preprocess, forward = build_model_and_preprocess()  # type: ignore

    sr = build_super_resolver()
    ds = Cityscapes(DATASET_ROOT, split="val", mode="fine", target_type="semantic")
    loader = DataLoader(ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=lambda b: tuple(zip(*b)))

    log: List[Dict[str,Any]] = []
    idx = 0

    for imgs, masks in tqdm(loader, desc="Inference", unit="batch"):
        if MODEL_BACKBONE == "segformer":
            logits_batch = forward(list(imgs))
        else:
            batch_t = torch.stack([preprocess(p) for p in imgs]).to(DEVICE)  # type: ignore[arg-type]
            logits_batch = forward(batch_t)

        for i in range(len(imgs)):
            img_np = np.array(imgs[i]); gt_np = np.vectorize(ID2TRAINID.get)(np.array(masks[i]))
            logits = logits_batch[i]
            if logits.shape[1:] != img_np.shape[:2]:
                logits = torch.nn.functional.interpolate(logits.unsqueeze(0), size=img_np.shape[:2], mode="bilinear", align_corners=False)[0]

            logits_np = (logits - logits.min()).div(logits.max()-logits.min()).cpu().numpy()
            max_prob = logits_np.max(0)

            # forward/high pred map
            high_mask = max_prob >= HIGH_CONF_THRESH
            logits_high = logits_np * high_mask
            high_pred = logits_high.argmax(0).astype(np.uint8)

            # reverse/low pred map
            low_mask = max_prob <= LOW_CONF_THRESH
            logits_low = logits_np * low_mask
            low_pred = logits_low.argmax(0).astype(np.uint8)

            closed = closing(high_pred)
            opened = opening(closed)
            refined = opened.copy()

            # small‑object SR refine
            for cid in SMALL_OBJECT_CLASSES:
                cls_m = (opened == cid).astype(np.uint8)
                contours,_ = cv2.findContours(cls_m, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                for cnt in contours:
                    x,y,w,h = cv2.boundingRect(cnt)
                    if w < MIN_INSTANCE_SIDE or h < MIN_INSTANCE_SIDE: continue
                    crop = crop_with_padding(img_np,x,y,w,h)
                    sr_crop = sr(crop)
                    if MODEL_BACKBONE == "segformer":
                        inp = extractor(images=Image.fromarray(sr_crop), return_tensors="pt").to(DEVICE)
                        with torch.no_grad(): crop_logits = net(**inp).logits
                        crop_logits = torch.nn.functional.interpolate(crop_logits, size=sr_crop.shape[:2], mode="bilinear", align_corners=False)[0]
                    else:
                        ct = T.ToTensor()(sr_crop).unsqueeze(0)
                        ct = T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])(ct).to(DEVICE)
                        with torch.no_grad(): crop_logits = net(ct)[0]
                    labels = torch.softmax(crop_logits,0).cpu().numpy().argmax(0)
                    labels = cv2.resize(labels,(w,h),interpolation=cv2.INTER_NEAREST)
                    refined[y:y+h, x:x+w] = labels

            raw_pred = logits_np.argmax(0).astype(np.uint8)

            log.append({
                "idx": idx,
                "iou_raw": calculate_iou_all(raw_pred, gt_np),
                "iou_high": calculate_iou_all(high_pred, gt_np),
                "iou_low": calculate_iou_all(low_pred, gt_np),
                "iou_closed": calculate_iou_all(closed, gt_np),
                "iou_opened": calculate_iou_all(opened, gt_np),
                "iou_refined": calculate_iou_all(refined, gt_np)
            })
            idx += 1

    RUN_PARAMS["finish_time"] = datetime.datetime.now().strftime("%Y-%m-%d_%H:%M:%S")
    with open(OUTPUT_JSON,"w",encoding="utf-8") as f:
        json.dump({"run_parameters": RUN_PARAMS, "metrics": log}, f, indent=2)

    print(f"Saved run → {OUTPUT_JSON}")

if __name__ == "__main__":
    main()
```
